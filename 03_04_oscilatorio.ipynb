# 3.4 Movimiento Oscilatorio de un ElectrĂłn

Consideremos un anillo de radio $R$ con una densidad de carga lineal constante $\lambda$. Queremos analizar el movimiento de un electrĂłn ubicado en el eje del anillo (eje $z$), a una distancia $z$ del centro, tal que $z \ll R$.

Primero, determinamos el potencial elĂ©ctrico en un punto sobre el eje $z$. Tomando un diferencial de carga $dq' = \lambda dl'$ a una distancia $\sqrt{z^2 + R^2}$ del punto de interĂ©s, integramos sobre todo el anillo:

$$V(z) = \frac{1}{4\pi\epsilon_0} \int \frac{\lambda dl'}{\sqrt{z^2 + R^2}} = \frac{1}{4\pi\epsilon_0} \frac{\lambda (2\pi R)}{\sqrt{z^2 + R^2}}$$

$$V(z) = \frac{\lambda R}{2\epsilon_0} (z^2 + R^2)^{-1/2}$$

La energĂ­a mecĂˇnica total del electrĂłn, que se conserva, es la suma de su energĂ­a cinĂ©tica y potencial ($U = qV = -eV$):

$$E = K + U = \frac{1}{2}m_e v^2 - e \frac{\lambda R}{2\epsilon_0} (z^2 + R^2)^{-1/2}$$

Dado que el electrĂłn se encuentra muy cerca del centro ($z \ll R$), podemos aproximar el tĂ©rmino $(z^2 + R^2)^{-1/2}$ utilizando una expansiĂłn de Taylor de primer orden:

$$(z^2 + R^2)^{-1/2} = \frac{1}{R} \left(1 + \frac{z^2}{R^2}\right)^{-1/2} \approx \frac{1}{R} \left(1 - \frac{z^2}{2R^2}\right)$$

Sustituyendo esta aproximaciĂłn geomĂ©trica en la ecuaciĂłn de la energĂ­a:

$$E \approx \frac{1}{2}m_e v^2 - \frac{e\lambda R}{2\epsilon_0} \frac{1}{R} \left(1 - \frac{z^2}{2R^2}\right) = \frac{1}{2}m_e v^2 - \frac{e\lambda}{2\epsilon_0} \left(1 - \frac{z^2}{2R^2}\right)$$

Como la fuerza es conservativa y la energĂ­a mecĂˇnica es constante, su derivada temporal es nula $\left(\frac{dE}{dt} = 0\right)$. Sabiendo que $v = \frac{dz}{dt}$ y la aceleraciĂłn es $\frac{dv}{dt} = \frac{d^2z}{dt^2}$:

$$\begin{aligned}\frac{1}{2}m_e \left(2v \frac{dv}{dt}\right) - \frac{e\lambda}{2\epsilon_0} \left( - \frac{2z}{2R^2} \frac{dz}{dt} \right) &= 0 \\
m_e v \frac{d^2z}{dt^2} + \frac{e\lambda}{2\epsilon_0 R^2} z v &= 0\end{aligned}$$

Dividiendo toda la expresiĂłn por la velocidad $v$ (asumiendo que no es permanentemente nula), llegamos a la ecuaciĂłn de movimiento para el electrĂłn:

$$m_e \frac{d^2z}{dt^2} + \frac{e\lambda}{2\epsilon_0 R^2} z = 0 \implies \frac{d^2z}{dt^2} + \left( \frac{e\lambda}{2\epsilon_0 R^2 m_e} \right) z = 0$$

Esta es exactamente la ecuaciĂłn diferencial de un oscilador armĂłnico simple $\left(\frac{d^2z}{dt^2} + \omega^2 z = 0\right)$, cuya soluciĂłn general tiene la forma $z(t) = A \cos(\omega t + \phi)$.

A partir de esto, identificamos la frecuencia angular de oscilaciĂłn $\omega$:

$$\omega = \sqrt{\frac{e\lambda}{2\epsilon_0 R^2 m_e}}$$

Y finalmente, el perĂ­odo de oscilaciĂłn $T = \frac{2\pi}{\omega}$ queda determinado por:

$$T = 2\pi \sqrt{\frac{2\epsilon_0 R^2 m_e}{e\lambda}}$$
