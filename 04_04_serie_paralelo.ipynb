# 4.4 Condensadores en Circuitos

En los circuitos eléctricos, los condensadores pueden combinarse de diversas formas. Nos enfocaremos en las dos configuraciones principales: paralelo y serie.

## Condensadores en Paralelo
Cuando los condensadores están conectados en paralelo, la diferencia de potencial es la misma para todos ellos, pero la carga total se distribuye.


&nbsp;
```{image} images_tikz/04_07.svg
:alt: Condensadores en paralelo
:width: 50%
:align: center
&nbsp;


* **Voltaje:** $V_{\text{eq}} = V_1 = V_2 = \dots = V_0$
* **Carga Total:** $Q_{\text{Total}} = Q_1 + Q_2 = C_1V_1 + C_2V_2$
* **Capacitancia Equivalente:** $C_{\text{eq}} = C_1 + C_2$

En general, para $N$ condensadores en paralelo:

$$C_{\text{eq}} = \sum_{i=1}^{N} C_i$$

## Condensadores en Serie
En una conexión en serie, la carga en todas las placas es idéntica por conservación de carga, pero el voltaje total se divide entre los condensadores.


&nbsp;
```{image} images_tikz/04_08.svg
:alt: Condensadores en serie
:width: 50%
:align: center
&nbsp;


* **Carga:** $Q_1 = Q_2 = Q_{\text{eq}} = Q_0$
* **Voltaje Total:** $V_0 = V_1 + V_2 = \frac{Q_1}{C_1} + \frac{Q_2}{C_2}$
* **Capacitancia Equivalente:** $\frac{1}{C_{\text{eq}}} = \frac{1}{C_1} + \frac{1}{C_2}$

En general, para $N$ condensadores en serie:

$$\frac{1}{C_{\text{eq}}} = \sum_{i=1}^{N} \frac{1}{C_i}$$



## Ejemplo: Análisis de un circuito con condensadores

Consideremos un circuito inicial con una fuente de voltaje $V_0$, dos interruptores ($S_1$ y $S_2$), y tres condensadores ($C_0$, $C_1$ y $C_2$).


&nbsp;
```{image} images_tikz/04_09.svg
:alt: Circuito de ejemplo
:width: 60%
:align: center
&nbsp;


Datos del problema:
* $C_1 = 2C_0$
* $C_2 = 6C_0$

**a) Cerramos $S_1$ manteniendo abierto $S_2$:**

Al cerrar $S_1$, el condensador $C_0$ se conecta directamente a la batería.


&nbsp;
```{image} images_tikz/04_10.svg
:alt: Interruptor 1 cerrado
:width: 40%
:align: center
&nbsp;


Una vez que se alcanza el equilibrio electrostático, la carga almacenada por el condensador $C_0$ es:

$$Q_0 = C_0 V_0$$

**b) Cerramos $S_2$ y abrimos $S_1$:**

Al abrir $S_1$, la fuente de voltaje queda completamente desconectada del sistema. Al cerrar $S_2$, el condensador $C_0$ (que ahora actúa como fuente aportando su carga inicial $Q_0$) queda conectado al resto del circuito.


&nbsp;
```{image} images_tikz/04_11.svg
:alt: Interruptor 2 cerrado
:width: 40%
:align: center
&nbsp;


Al observar la nueva configuración, notamos que:
* $C_1$ está en serie con $C_2$.
* El conjunto $C_1-C_2$ está en paralelo con $C_0$.


&nbsp;
```{image} images_tikz/04_12.svg
:alt: Interruptor 2 cerrado
:width: 40%
:align: center
&nbsp;


Para resolver cómo se distribuye la carga, primero calculamos la capacitancia equivalente de la rama en serie ($C_{1,2}$):

$$\frac{1}{C_{1,2}} = \frac{1}{C_1} + \frac{1}{C_2} = \frac{1}{2C_0} + \frac{1}{6C_0}$$
$$C_{1,2} = \frac{3}{2}C_0$$

Ahora, nuestro circuito equivalente consiste en $C_0$ en paralelo con $C_{1,2}$. La capacitancia equivalente total del sistema ($C_{eq}$) es:

$$C_{eq} = C_0 + C_{1,2} = C_0 + \frac{3}{2}C_0 = \frac{5}{2}C_0$$

Por el principio de conservación de la carga, la carga total del circuito sigue siendo $Q_0$. Esta carga se distribuye en la nueva capacitancia equivalente. El nuevo voltaje común entre los nodos en paralelo (llamémoslo $\Delta V_{ab}$) es:

$$Q_0 = C_{eq} \Delta V_{ab} = \frac{5}{2}C_0 \cdot \Delta V_{ab}$$
$$\Delta V_{ab} = \frac{2}{5} \frac{Q_0}{C_0}$$

**Después de la distribución de la carga:**

Conociendo el nuevo voltaje del sistema, calculamos cómo se repartió la carga original. 

Para el condensador $C_0$:

$$Q'_0 = C_0 \Delta V_{ab} = C_0 \left( \frac{2}{5} \frac{Q_0}{C_0} \right) \implies Q'_0 = \frac{2}{5}Q_0$$

Para la rama equivalente $C_{1,2}$:

$$Q_{1,2} = C_{1,2} \Delta V_{ab} = \left( \frac{3}{2}C_0 \right) \left( \frac{2}{5} \frac{Q_0}{C_0} \right) \implies Q_{1,2} = \frac{3}{5}Q_0$$

Dado que los condensadores $C_1$ y $C_2$ están en serie, la carga en cada uno de ellos es idéntica a la carga de su capacitancia equivalente:

$$Q'_1 = Q'_2 = Q_{1,2} = \frac{3}{5}Q_0$$

Finalmente, podemos determinar las nuevas diferencias de potencial individuales para $C_1$ y $C_2$:

$$V'_1 = \frac{Q'_1}{C_1} = \frac{\frac{3}{5}Q_0}{2C_0} = \frac{3}{10}\frac{Q_0}{C_0}$$
$$V'_2 = \frac{Q'_2}{C_2} = \frac{\frac{3}{5}Q_0}{6C_0} = \frac{3}{30}\frac{Q_0}{C_0} = \frac{1}{10}\frac{Q_0}{C_0}$$
