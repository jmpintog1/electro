# Capítulo 1: Carga eléctrica y campo eléctrico

La **carga eléctrica** ($q$) es una propiedad intrínseca de la materia que determina la intensidad de las interacciones electromagnéticas.

### Naturaleza de la carga
* **Protones:** Poseen carga positiva ($q > 0$).
* **Electrones:** Poseen carga negativa ($q < 0$).
* **Neutrones:** Son eléctricamente neutros ($q = 0$).

### Unidades y Cuantización
La unidad de medida de la carga eléctrica en el Sistema Internacional es el **Coulomb** ($C$).

La **carga elemental** representa la magnitud de carga más pequeña que puede existir de forma aislada:
$e = 1{,}6 \times 10^{-19} \text{ C}$

De acuerdo a esto, las partículas fundamentales se definen como:
* **Protones:** $q_p = +e$
* **Electrones:** $q_e = -e$

> **Nota:** Un cuerpo adquiere carga neta al ceder o ganar electrones. Si pierde electrones, resulta con carga positiva; si los gana, su carga será negativa.

# Ley de Coulomb

La interacción entre cargas eléctricas se rige por los siguientes principios fundamentales:

* **Tipos de carga:** Existen dos tipos de cargas, positivas y negativas.
* **Interacción a distancia:** Dos cargas interactúan con una fuerza cuya magnitud es inversamente proporcional al cuadrado de la distancia $d$ que las separa.
* **Proporcionalidad:** Esta fuerza es directamente proporcional al producto de las magnitudes de las cargas ($q_1, q_2$).

### Comportamiento de las fuerzas
* **Cargas de igual signo:** Se repelen. La fuerza $\vec{F}_{21}$ apunta hacia afuera de la línea que une las cargas.
* **Cargas de distinto signo:** Se atraen. La fuerza $\vec{F}_{21}$ apunta hacia la carga opuesta.

---

## Formulación Matemática (Escalar)

La magnitud de la fuerza de Coulomb se define como:

$$F = k \frac{q_1 q_2}{d^2} = \frac{1}{4\pi\epsilon_0} \frac{q_1 q_2}{d^2}$$

Donde las constantes fundamentales son:
* **Constante de Coulomb ($k$):** $k =  8{,}98755 \times 10^9 \ \frac{\text{N}\cdot\text{m}^2}{\text{C}^2} \approx 9 \times 10^9 \ \frac{\text{N}\cdot\text{m}^2}{\text{C}^2}$
* **Permitividad del vacío ($\epsilon_0$):** $\epsilon_0 = 8{,}854 \times 10^{-12} \ \frac{\text{C}^2}{\text{N}\cdot\text{m}^2}$

---

## Formulación Vectorial

Para un análisis preciso en el espacio, definimos la fuerza que la carga $2$ ejerce sobre la carga $1$ ($\vec{F}_{12}$) utilizando vectores de posición $\vec{r}_1$ y $\vec{r}_2$:

$$
\begin{aligned}
\vec{F}_{12} &= \frac{1}{4\pi\epsilon_0} \frac{q_1 q_2}{d^2} \hat{u} \\
&= \frac{1}{4\pi\epsilon_0} \frac{q_1 q_2}{|\vec{r}_1 - \vec{r}_2|^2} \frac{(\vec{r}_1 - \vec{r}_2)}{|\vec{r}_1 - \vec{r}_2|} \\
\vec{F}_{12} &= \frac{1}{4\pi\epsilon_0} q_1 q_2 \frac{(\vec{r}_1 - \vec{r}_2)}{|\vec{r}_1 - \vec{r}_2|^3}
\end{aligned}
$$

## Principio de Superposición (Cargas Discretas)
Si tenemos un sistema de $N$ cargas puntuales, la fuerza total resultante sobre una carga $q_1$ es la suma vectorial de las fuerzas individuales ejercidas por el resto de las cargas:

$$\vec{F}_1 = \vec{F}_{12} + \vec{F}_{13} + \dots + \vec{F}_{1N}$$

En términos generales:

$$\vec{F}_1 = \frac{q_1}{4\pi\epsilon_0} \sum_{j=2}^{N} q_j \frac{\vec{r}_1 - \vec{r}_j}{|\vec{r}_1 - \vec{r}_j|^3}$$

> **Nota:** Este análisis es válido estrictamente para distribuciones de **cargas puntuales**.

````python
import numpy as np
import matplotlib.pyplot as plt

# --- PARÁMETROS QUE EL ESTUDIANTE PUEDE MODIFICAR ---
# Carga 1 (en el origen)
q1 = 1.0   # Carga positiva
x1, y1 = -1.0, 0.0

# Carga 2 (puedes cambiarla a positiva para ver repulsión)
q2 = -1.0  # Carga negativa (Dipolo)
x2, y2 = 1.0, 0.0
# ----------------------------------------------------

def campo_electrico(x, y, q, x_c, y_c):
    k = 8.99e9
    # Distancia en componentes
    dx = x - x_c
    dy = y - y_c
    # Distancia al cuadrado y al cubo (para el vector unitario)
    r2 = dx**2 + dy**2
    r3 = r2**(1.5)
    
    # Evitar división por cero (pequeño truco numérico)
    r3[r3 == 0] = 1e-10
    
    # Componentes del campo E = k * q / r^2 * (r_vec / r) = k * q * r_vec / r^3
    Ex = k * q * dx / r3
    Ey = k * q * dy / r3
    return Ex, Ey

# Crear una malla de puntos (el "espacio")
x = np.linspace(-3, 3, 100)
y = np.linspace(-2, 2, 100)
X, Y = np.meshgrid(x, y)

# Calcular campo total (Principio de Superposición)
Ex1, Ey1 = campo_electrico(X, Y, q1, x1, y1)
Ex2, Ey2 = campo_electrico(X, Y, q2, x2, y2)

Ex_total = Ex1 + Ex2
Ey_total = Ey1 + Ey2
E_magnitud = np.sqrt(Ex_total**2 + Ey_total**2)

# --- GRAFICAR (La parte visual "Rupturista") ---
plt.figure(figsize=(10, 6))

# Mapa de color logarítmico para la intensidad (para ver el decaimiento)
# Usamos log porque cerca de la carga el campo tiende a infinito
plt.pcolormesh(X, Y, np.log10(E_magnitud), cmap='inferno', shading='auto', alpha=0.3)

# Líneas de campo (Streamplot)
plt.streamplot(X, Y, Ex_total, Ey_total, color='k', linewidth=1, density=1.5, arrowstyle='->')

# Dibujar las cargas
plt.scatter([x1, x2], [y1, y2], c=['red' if q1>0 else 'blue', 'red' if q2>0 else 'blue'], s=200, zorder=10, edgecolors='black')
plt.text(x1, y1+0.15, f"$q_1={q1}$", ha='center', fontsize=12, fontweight='bold')
plt.text(x2, y2+0.15, f"$q_2={q2}$", ha='center', fontsize=12, fontweight='bold')

plt.title(r'Campo Eléctrico Total: $\vec{E} = \vec{E}_1 + \vec{E}_2$', fontsize=14)
plt.xlabel('Distancia x (m)')
plt.ylabel('Distancia y (m)')
plt.xlim(-3, 3)
plt.ylim(-2, 2)
plt.grid(True, linestyle='--', alpha=0.6)

plt.show()
````

````python
import numpy as np
import plotly.graph_objects as go
import plotly.figure_factory as ff
from IPython.display import HTML

# --- 1. CONFIGURACIÓN DEL ESPACIO (Malla de puntos) ---
# Usamos menos puntos que en el mapa de calor para que las flechas no se amontonen
N = 25
x_range = np.linspace(-3, 3, N)
y_range = np.linspace(-2, 2, N)
X, Y = np.meshgrid(x_range, y_range)

# Posiciones de las cargas
x1, y1 = -1.0, 0.0 # Carga fija Q1
x2, y2 = 1.0, 0.0  # Carga variable Q2

# --- 2. FUNCIÓN DE CÁLCULO VECTORIAL ---
def calcular_vectores_campo(q_variable):
    k = 8.99e9
    epsilon = 0.1 # Factor de suavizado para evitar flechas infinitas cerca de las cargas

    # Vector r1 (desde Q1 hasta cada punto de la malla)
    dx1 = X - x1
    dy1 = Y - y1
    r1_sq = dx1**2 + dy1**2 + epsilon**2 # Distancia suavizada al cuadrado
    r1 = np.sqrt(r1_sq)
    # Campo E1 (Componentes vectoriales E = k*q*vec_r / r^3)
    Ex1 = k * 1.0 * dx1 / (r1**3)
    Ey1 = k * 1.0 * dy1 / (r1**3)

    # Vector r2 (desde Q2 variable)
    dx2 = X - x2
    dy2 = Y - y2
    r2_sq = dx2**2 + dy2**2 + epsilon**2
    r2 = np.sqrt(r2_sq)
    # Campo E2
    Ex2 = k * q_variable * dx2 / (r2**3)
    Ey2 = k * q_variable * dy2 / (r2**3)

    # Principio de Superposición (Suma vectorial)
    Ex_total = Ex1 + Ex2
    Ey_total = Ey1 + Ey2
    
    return Ex_total, Ey_total

# --- 3. GENERACIÓN DE FRAMES (VISUALIZACIÓN) ---
# Valores de carga Q2 para el deslizador
valores_q2 = np.linspace(-5, 5, 21) # 21 pasos para una animación fluida

# --- Función auxiliar para crear la figura de flechas ---
def crear_figura_quiver(q_val):
    Ex, Ey = calcular_vectores_campo(q_val)
    # Usamos figure_factory para crear el mapa de flechas (Quiver)
    # scale: ajusta la longitud de las flechas
    # arrow_scale: ajusta el tamaño de la cabeza de la flecha
    fig_quiver = ff.create_quiver(X, Y, Ex, Ey, scale=0.02, arrow_scale=0.3, name='Líneas de Campo', line=dict(color='black', width=1))
    return fig_quiver

# Creamos la FIGURA BASE (Estado inicial)
fig_base = crear_figura_quiver(valores_q2[0])
fig = go.Figure(data=fig_base.data)

# Creamos los FRAMES de animación
frames = []
for q_val in valores_q2:
    # Generamos una figura temporal para este valor de Q
    temp_fig = crear_figura_quiver(q_val)
    # Extraemos los datos de las líneas de esa figura y creamos un frame
    frames.append(go.Frame(data=temp_fig.data, name=f"q={q_val:.1f}"))

fig.frames = frames

# --- 4. CONFIGURACIÓN DEL LAYOUT (Deslizador y Cargas) ---
# Añadir marcadores para las cargas
fig.add_trace(go.Scatter(x=[x1], y=[y1], mode='markers+text', marker=dict(size=18, color='red', line=dict(width=2, color='black')), name='Q1 (+1C)', text=['+'], textposition='middle center', textfont=dict(color='white', size=14)))
fig.add_trace(go.Scatter(x=[x2], y=[y2], mode='markers+text', marker=dict(size=18, color='blue', line=dict(width=2, color='black')), name='Q2 (Variable)', text=['Q2'], textposition='top center'))

# Configurar el deslizador
sliders = [dict(
    active=0,
    currentvalue={"prefix": "Carga Q2: ", "suffix": " C"},
    pad={"t": 50},
    steps=[dict(
        method='animate',
        # args: [nombre_frame, configuración_animación]
        args=[[f"q={qv:.1f}"], dict(mode='immediate', frame=dict(duration=0, redraw=False), transition=dict(duration=0))],
        label=f"{qv:.1f}"
    ) for qv in valores_q2]
)]

fig.update_layout(
    title="Dirección del Campo Eléctrico (Diagrama Vectorial)",
    xaxis=dict(title="Distancia X", range=[-3.2, 3.2], zeroline=False),
    yaxis=dict(title="Distancia Y", range=[-2.2, 2.2], zeroline=False),
    sliders=sliders,
    width=750, height=550,
    showlegend=False,
    plot_bgcolor='rgba(240,240,240,0.5)' # Fondo gris claro
)

# --- 5. INYECCIÓN HTML PARA LA WEB ---
HTML(fig.to_html(include_plotlyjs='cdn'))
````

````python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.patches import Circle

# 1. Configuración del sistema
d = 4.0               # Distancia arbitraria entre cargas
q1_mag = 1.0          # Carga +q
q2_mag = 2.0          # Carga +2q
x1, y1 = 0, 0         # Posición de q1
x2, y2 = d, 0         # Posición de q2

# 2. Creación de la malla (Grid) para calcular el campo
# Creamos un espacio alrededor de las cargas
x = np.linspace(-d/2, d*1.5, 200)
y = np.linspace(-d, d, 200)
X, Y = np.meshgrid(x, y)

# 3. Cálculo del Campo Eléctrico (Principio de Superposición)
# E_total = E1 + E2
# E = k * q / r^2 * r_hat

# Contribución de q1
dx1 = X - x1
dy1 = Y - y1
r1 = np.sqrt(dx1**2 + dy1**2)
# Evitamos división por cero en la posición exacta de la carga
r1[r1 < 0.1] = 0.1 
Ex1 = (q1_mag / r1**2) * (dx1 / r1)
Ey1 = (q1_mag / r1**2) * (dy1 / r1)

# Contribución de q2
dx2 = X - x2
dy2 = Y - y2
r2 = np.sqrt(dx2**2 + dy2**2)
r2[r2 < 0.1] = 0.1
Ex2 = (q2_mag / r2**2) * (dx2 / r2)
Ey2 = (q2_mag / r2**2) * (dy2 / r2)

# Campo Neto
Ex = Ex1 + Ex2
Ey = Ey1 + Ey2
E_magnitud = np.sqrt(Ex**2 + Ey**2)

# 4. Graficar
fig, ax = plt.subplots(figsize=(10, 6))

# Dibujar líneas de campo (Streamplot)
# Usamos log de la magnitud para el color, para resaltar zonas intensas
color_map = np.log(E_magnitud)
strm = ax.streamplot(X, Y, Ex, Ey, color=color_map, linewidth=1, 
                     cmap='inferno', density=1.5, arrowstyle='->', arrowsize=1.5)

# Dibujar las cargas como círculos
c1 = Circle((x1, y1), 0.2, color='blue', alpha=0.8, zorder=10)
c2 = Circle((x2, y2), 0.3, color='red', alpha=0.8, zorder=10) # Más grande pq es 2q
ax.add_patch(c1)
ax.add_patch(c2)

# Etiquetas de las cargas
ax.text(x1, y1-0.5, r'$+q$', fontsize=14, ha='center', va='top', color='blue', weight='bold')
ax.text(x2, y2-0.5, r'$+2q$', fontsize=14, ha='center', va='top', color='red', weight='bold')

# --- EL PUNTO CLAVE: EL EQUILIBRIO ---
# Calculamos la posición teórica derivada en el texto
# x_eq = (sqrt(2) - 1) * d
x_eq = (np.sqrt(2) - 1) * d
y_eq = 0

# Marcamos el punto de equilibrio
ax.plot(x_eq, y_eq, 'X', color='lime', markersize=12, markeredgecolor='black', zorder=15, label='Punto de Equilibrio')

# Configuración final del gráfico
ax.set_title(f'Campo Eléctrico Resultante\nEquilibrio teórico en x = {x_eq/d:.3f}d', fontsize=14)
ax.set_xlabel('Distancia (x)')
ax.set_ylabel('Distancia (y)')
ax.set_aspect('equal')
ax.legend(loc='upper right')
ax.grid(True, linestyle='--', alpha=0.3)

# Ocultar código en Jupyter Book (opcional, si usas etiquetas de celda)
plt.tight_layout()
plt.show()
````
