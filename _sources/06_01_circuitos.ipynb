# 6.1 Simbología y Conceptos Básicos

Para analizar circuitos eléctricos, primero debemos definir la simbología estándar de los componentes:
* **Batería (Fuente de fem $\mathcal{E}$):** Provee la diferencia de potencial.
* **Condensador (Capacitor):** Almacena carga eléctrica.
* **Resistencia:** Representa la oposición al flujo de corriente.
* **Amperímetro (A):** Mide la corriente eléctrica.
* **Voltímetro (V):** Mide la diferencia de potencial.

En el circuito más simple (una batería conectada a una resistencia), la diferencia de potencial $V$ entregada por la batería depende de si es ideal o real:
* **Batería ideal:** $V = \mathcal{E}$
* **Batería no ideal:** $V = \mathcal{E} - Ir$ (donde $r$ es la resistencia interna de la batería).

# 6.2 Potencia Eléctrica

La energía $U$ entregada a una carga $q$ a través de un potencial $V$ está dada por $dU = dq \cdot V$. Si derivamos esto con respecto al tiempo, obtenemos la rapidez con la que se transfiere la energía, es decir, la **Potencia** ($P$):

$$P = \frac{dU}{dt} = \frac{dq}{dt} V \implies P = IV$$

La unidad de medida de la potencia es el Watt ($W$), que equivale a un Joule por segundo ($1 \, \text{W} = 1 \, \text{J/s}$).
Para el caso específico de una resistencia, usando la Ley de Ohm ($V = IR$), la potencia disipada se puede expresar como:

$$P = I^2 R$$

# 6.3 Conexiones entre Resistencias

## Conexión en Serie

Cuando las resistencias se conectan una tras otra, la corriente que fluye a través de ellas es la misma, pero el voltaje total se divide entre cada componente:
* Corriente: $I = I_1 = I_2$
* Voltaje: $V = V_1 + V_2$

Aplicando la Ley de Ohm ($V = I_1R_1 + I_2R_2$), obtenemos que la resistencia equivalente ($R_{eq}$) es la suma algebraica de las resistencias individuales:

$$R_{eq} = \sum_{j=1}^{N} R_j$$

## Conexión en Paralelo

En una conexión en paralelo, los terminales de las resistencias están conectados a los mismos puntos, por lo que la diferencia de potencial es la misma para todas, pero la corriente total se divide:
* Voltaje: $V = V_1 = V_2$
* Corriente: $I = I_1 + I_2$

Al sustituir las corrientes ($I_1 = V/R_1$ y $I_2 = V/R_2$), la resistencia equivalente se calcula mediante la suma de sus inversos:

$$\frac{1}{R_{eq}} = \sum_{j=1}^{N} \frac{1}{R_j}$$

# 6.4 Reglas de Kirchhoff

Para circuitos más complejos donde no es posible simplificar únicamente con equivalencias en serie y paralelo, utilizamos las reglas de Kirchhoff.

## Regla de los Nodos (Conservación de la Carga)

En cualquier nodo (punto de unión de tres o más conductores), la suma de las corrientes que entran debe ser igual a la suma de las corrientes que salen. Esto implica que no hay acumulación de carga:

$$\sum_{i=1}^{N} I_i = 0$$

## Regla de las Mallas (Conservación de la Energía)

La suma algebraica de las diferencias de potencial en cualquier trayectoria cerrada (malla) de un circuito debe ser cero:

$$\sum_{i=1}^{N} \mathcal{E}_i - \sum_{j=1}^{N} I_j R_j = 0$$
