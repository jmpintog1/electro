# 4.4 Condensadores en Circuitos

En los circuitos eléctricos, los condensadores pueden combinarse de diversas formas. Nos enfocaremos en las dos configuraciones principales: paralelo y serie.

## Condensadores en Paralelo
Cuando los condensadores están conectados en paralelo, la diferencia de potencial es la misma para todos ellos, pero la carga total se distribuye.


&nbsp;
```{image} images_tikz/04_07.svg
:alt: Condensadores en paralelo
:width: 50%
:align: center
```
&nbsp;


* **Voltaje:** $V_{\text{eq}} = V_1 = V_2 = \dots = V_0$
* **Carga Total:** $Q_{\text{Total}} = Q_1 + Q_2 = C_1V_1 + C_2V_2$
* **Capacitancia Equivalente:** $C_{\text{eq}} = C_1 + C_2$

En general, para $N$ condensadores en paralelo:

$$C_{\text{eq}} = \sum_{i=1}^{N} C_i$$

## Condensadores en Serie
En una conexión en serie, la carga en todas las placas es idéntica por conservación de carga, pero el voltaje total se divide entre los condensadores.


&nbsp;
```{image} images_tikz/04_08.svg
:alt: Condensadores en serie
:width: 50%
:align: center
```
&nbsp;


* **Carga:** $Q_1 = Q_2 = Q_{\text{eq}} = Q_0$
* **Voltaje Total:** $V_0 = V_1 + V_2 = \frac{Q_1}{C_1} + \frac{Q_2}{C_2}$
* **Capacitancia Equivalente:** $\frac{1}{C_{\text{eq}}} = \frac{1}{C_1} + \frac{1}{C_2}$

En general, para $N$ condensadores en serie:

$$\frac{1}{C_{\text{eq}}} = \sum_{i=1}^{N} \frac{1}{C_i}$$


## Ejemplo: Análisis de un circuito con condensadores

Consideremos un circuito inicial con una fuente de voltaje $V_0$, dos interruptores ($S_1$ y $S_2$), y tres condensadores ($C_0$, $C_1$ y $C_2$).


&nbsp;
```{image} images_tikz/04_09.svg
:alt: Circuito de ejemplo
:width: 60%
:align: center
```
&nbsp;


Datos del problema:
* $C_1 = 2C_0$
* $C_2 = 6C_0$

**a) Cerramos $S_1$ manteniendo abierto $S_2$:**

Al cerrar $S_1$, el condensador $C_0$ se conecta directamente a la batería.


&nbsp;
```{image} images_tikz/04_10.svg
:alt: Interruptor 1 cerrado
:width: 40%
:align: center
```
&nbsp;


Una vez que se alcanza el equilibrio electrostático, la carga almacenada por el condensador $C_0$ es:

$$Q_0 = C_0 V_0$$

**b) Cerramos $S_2$ y abrimos $S_1$:**

Al abrir $S_1$, la fuente de voltaje queda completamente desconectada del sistema. Al cerrar $S_2$, el condensador $C_0$ (que ahora actúa como fuente aportando su carga inicial $Q_0$) queda conectado al resto del circuito.


&nbsp;
```{image} images_tikz/04_11.svg
:alt: Interruptor 2 cerrado
:width: 40%
:align: center
```
&nbsp;


Al observar la nueva configuración, notamos que:
* $C_1$ está en serie con $C_2$.
* El conjunto $C_1-C_2$ está en paralelo con $C_0$.


&nbsp;
```{image} images_tikz/04_12.svg
:alt: Interruptor 2 cerrado
:width: 40%
:align: center
```
&nbsp;


Para resolver cómo se distribuye la carga, primero calculamos la capacitancia equivalente de la rama en serie ($C_{1,2}$):

$$\frac{1}{C_{1,2}} = \frac{1}{C_1} + \frac{1}{C_2} = \frac{1}{2C_0} + \frac{1}{6C_0}$$
$$C_{1,2} = \frac{3}{2}C_0$$

Ahora, nuestro circuito equivalente consiste en $C_0$ en paralelo con $C_{1,2}$. La capacitancia equivalente total del sistema ($C_{eq}$) es:

$$C_{eq} = C_0 + C_{1,2} = C_0 + \frac{3}{2}C_0 = \frac{5}{2}C_0$$

Por el principio de conservación de la carga, la carga total del circuito sigue siendo $Q_0$. Esta carga se distribuye en la nueva capacitancia equivalente. El nuevo voltaje común entre los nodos en paralelo (llamémoslo $\Delta V_{ab}$) es:

$$Q_0 = C_{eq} \Delta V_{ab} = \frac{5}{2}C_0 \cdot \Delta V_{ab}$$
$$\Delta V_{ab} = \frac{2}{5} \frac{Q_0}{C_0}$$

**Después de la distribución de la carga:**

Conociendo el nuevo voltaje del sistema, calculamos cómo se repartió la carga original. 

Para el condensador $C_0$:

$$Q'_0 = C_0 \Delta V_{ab} = C_0 \left( \frac{2}{5} \frac{Q_0}{C_0} \right) \implies Q'_0 = \frac{2}{5}Q_0$$

Para la rama equivalente $C_{1,2}$:

$$Q_{1,2} = C_{1,2} \Delta V_{ab} = \left( \frac{3}{2}C_0 \right) \left( \frac{2}{5} \frac{Q_0}{C_0} \right) \implies Q_{1,2} = \frac{3}{5}Q_0$$

Dado que los condensadores $C_1$ y $C_2$ están en serie, la carga en cada uno de ellos es idéntica a la carga de su capacitancia equivalente:

$$Q'_1 = Q'_2 = Q_{1,2} = \frac{3}{5}Q_0$$

Finalmente, podemos determinar las nuevas diferencias de potencial individuales para $C_1$ y $C_2$:

$$V'_1 = \frac{Q'_1}{C_1} = \frac{\frac{3}{5}Q_0}{2C_0} = \frac{3}{10}\frac{Q_0}{C_0}$$
$$V'_2 = \frac{Q'_2}{C_2} = \frac{\frac{3}{5}Q_0}{6C_0} = \frac{3}{30}\frac{Q_0}{C_0} = \frac{1}{10}\frac{Q_0}{C_0}$$

````python
from IPython.display import display, HTML

simulacion_analisis_final_html = """
<div style="border: 1px solid #e0e0e0; padding: 20px; border-radius: 8px; background-color: #f8f9fa; font-family: -apple-system, BlinkMacSystemFont, 'Segoe UI', Roboto, Arial, sans-serif;">
    <h3 style="margin-top:0; color: #2c3e50;">Simulación 2D: Análisis de Circuito (Conexiones Perfectas)</h3>
    <p style="font-size: 0.95em; color: #555; margin-bottom: 15px;">Sigue la secuencia cerrando y abriendo los interruptores para observar el principio de conservación de la carga.</p>
    
    <div style="display: flex; gap: 20px; flex-wrap: wrap; margin-bottom: 15px;">
        <div style="flex: 1; min-width: 250px;">
            <div style="margin-bottom: 15px;">
                <label style="font-weight: 600; display: inline-block; width: 140px;">Voltaje V₀ (V): <span id="ccaf_val_v0" style="font-weight: normal;">10</span></label>
                <input type="range" id="ccaf_slider_v0" min="5" max="24" step="1" value="10" style="width: 50%; vertical-align: middle;">
            </div>
            <div style="margin-bottom: 15px;">
                <label style="font-weight: 600; display: inline-block; width: 140px;">Base C₀ (μF): <span id="ccaf_val_c0" style="font-weight: normal;">2</span></label>
                <input type="range" id="ccaf_slider_c0" min="1" max="10" step="1" value="2" style="width: 50%; vertical-align: middle;">
            </div>
            
            <div style="background: #e9ecef; padding: 10px; border-radius: 6px; margin-bottom: 10px;">
                <b style="display: block; margin-bottom: 5px;">Controles del Sistema:</b>
                <div style="margin-bottom: 5px;">
                    <label style="cursor: pointer; font-weight: bold; color: #dc3545;">
                        <input type="radio" name="ccaf_estado" value="0" checked> Estado 0: Todo abierto
                    </label>
                </div>
                <div style="margin-bottom: 5px;">
                    <label style="cursor: pointer; font-weight: bold; color: #198754;">
                        <input type="radio" name="ccaf_estado" value="1"> Fase a): Cerrar S₁ (Cargar C₀)
                    </label>
                </div>
                <div>
                    <label style="cursor: pointer; font-weight: bold; color: #0d6efd;">
                        <input type="radio" name="ccaf_estado" value="2"> Fase b): Abrir S₁, Cerrar S₂ (Distribuir)
                    </label>
                </div>
            </div>
        </div>

        <div style="flex: 1.2; min-width: 300px; font-size: 0.9em;">
            <div style="display: flex; gap: 10px;">
                <div style="flex: 1; background: #fff; padding: 10px; border-radius: 6px; border: 1px solid #ccc;">
                    <h4 style="margin: 0 0 5px 0;">Condensador C₀</h4>
                    <div><b>C₀ =</b> <span id="ccaf_out_c0">2.0</span> μF</div>
                    <div><b>V₀ =</b> <span id="ccaf_out_vc0">0.0</span> V</div>
                    <div><b>Q₀ =</b> <span id="ccaf_out_qc0" style="color: #198754; font-weight: bold;">0.0</span> μC</div>
                </div>
                <div style="flex: 1; background: #fff; padding: 10px; border-radius: 6px; border: 1px solid #ccc;">
                    <h4 style="margin: 0 0 5px 0;">Rama C₁ y C₂</h4>
                    <div><b>C₁ = 2C₀ =</b> <span id="ccaf_out_c1">4.0</span> μF</div>
                    <div><b>C₂ = 6C₀ =</b> <span id="ccaf_out_c2">12.0</span> μF</div>
                    <hr style="margin: 5px 0;">
                    <div><b>V₁ =</b> <span id="ccaf_out_vc1">0.0</span> V</div>
                    <div><b>V₂ =</b> <span id="ccaf_out_vc2">0.0</span> V</div>
                    <div><b>Q₁ = Q₂ =</b> <span id="ccaf_out_q12" style="color: #0d6efd; font-weight: bold;">0.0</span> μC</div>
                </div>
            </div>
            <div style="margin-top: 10px; background: #fff; padding: 10px; border-radius: 6px; border: 1px solid #ffc107; text-align: center;">
                <b>Carga Total del Sistema (Q_total):</b> <span id="ccaf_out_qtot" style="font-size: 1.1em; font-weight: bold;">0.0</span> μC
            </div>
        </div>
    </div>
    
    <div style="text-align: center; position: relative;">
        <canvas id="ccaf_canvas" width="650" height="250" style="background: white; border: 1px solid #ccc; border-radius: 4px; box-shadow: 0 2px 5px rgba(0,0,0,0.05); max-width: 100%;"></canvas>
    </div>
</div>

<script>
    (function() {
        const canvas = document.getElementById('ccaf_canvas');
        const ctx = canvas.getContext('2d');
        
        const slider_v0 = document.getElementById('ccaf_slider_v0');
        const slider_c0 = document.getElementById('ccaf_slider_c0');
        const radios_estado = document.getElementsByName('ccaf_estado');
        
        const val_v0 = document.getElementById('ccaf_val_v0');
        const val_c0 = document.getElementById('ccaf_val_c0');
        
        const out_c0 = document.getElementById('ccaf_out_c0');
        const out_vc0 = document.getElementById('ccaf_out_vc0');
        const out_qc0 = document.getElementById('ccaf_out_qc0');
        
        const out_c1 = document.getElementById('ccaf_out_c1');
        const out_c2 = document.getElementById('ccaf_out_c2');
        const out_vc1 = document.getElementById('ccaf_out_vc1');
        const out_vc2 = document.getElementById('ccaf_out_vc2');
        const out_q12 = document.getElementById('ccaf_out_q12');
        const out_qtot = document.getElementById('ccaf_out_qtot');

        // Geometría del circuito en el Eje X
        const x_bat = 80;
        const x_s1_left = 140;
        const x_s1_right = 180;
        const x_c0 = 280;
        const x_s2_left = 380;
        const x_s2_right = 420;
        const x_c12 = 520;
        
        // Geometría del circuito en el Eje Y
        const y_top = 50;
        const y_bot = 200;
        
        // Coordenadas Y centrales de los condensadores
        const y_c0 = 120;
        const y_c1 = 90;
        const y_c2 = 160;

        function drawSwitch(x_left, x_right, y, label, closed) {
            ctx.fillStyle = '#000';
            ctx.font = 'bold 14px Arial';
            ctx.textAlign = 'center';
            ctx.fillText(label, (x_left + x_right)/2, y - 25);
            
            ctx.beginPath();
            ctx.fillStyle = '#000';
            ctx.arc(x_left, y, 4, 0, Math.PI*2);
            ctx.arc(x_right, y, 4, 0, Math.PI*2);
            ctx.fill();
            
            ctx.beginPath();
            ctx.lineWidth = 3;
            ctx.strokeStyle = closed ? '#198754' : '#dc3545';
            ctx.moveTo(x_left, y);
            
            if(closed) {
                ctx.lineTo(x_right, y);
            } else {
                ctx.lineTo(x_left + (x_right - x_left)*0.8, y - 18);
            }
            ctx.stroke();
            ctx.textAlign = 'left';
        }

        function drawCap(x, y, label, Q, max_Q, color_rgb) {
            let w = 40; 
            let d = 6;   // Grosor de la placa
            let gap = 6; // Distancia desde el centro al borde interno de la placa
            
            // Coordenadas exactas de dibujo
            let y_top_plate = y - gap - d; // Borde superior exterior
            let y_bot_plate = y + gap;     // Borde inferior exterior

            let intensity = max_Q > 0 ? Math.min(1.0, Math.max(0.1, Q / max_Q)) : 0.1;
            if (Q === 0) intensity = 0;

            ctx.fillStyle = `rgba(${color_rgb}, ${intensity})`;
            ctx.fillRect(x - w/2, y_top_plate, w, d);
            ctx.fillRect(x - w/2, y_bot_plate, w, d);

            ctx.fillStyle = '#000';
            ctx.lineWidth = 1;
            ctx.strokeRect(x - w/2, y_top_plate, w, d);
            ctx.strokeRect(x - w/2, y_bot_plate, w, d);

            ctx.fillStyle = '#000';
            ctx.font = 'bold 14px Arial';
            ctx.fillText(label, x + 25, y + 5);
        }

        function updateSim() {
            let V0 = parseFloat(slider_v0.value);
            let C0 = parseFloat(slider_c0.value);
            let estado = parseInt(Array.from(radios_estado).find(r => r.checked).value);
            
            val_v0.innerText = V0;
            val_c0.innerText = C0;

            let C1 = 2 * C0;
            let C2 = 6 * C0;
            
            out_c0.innerText = C0.toFixed(1);
            out_c1.innerText = C1.toFixed(1);
            out_c2.innerText = C2.toFixed(1);

            let Q0_max = C0 * V0;
            let v_c0 = 0, q_c0 = 0, v_c1 = 0, v_c2 = 0, q_c12 = 0, q_total = 0;

            if (estado === 1) {
                v_c0 = V0; q_c0 = Q0_max; q_total = q_c0;
            } else if (estado === 2) {
                let C12_eq = (C1 * C2) / (C1 + C2);
                let C_tot = C0 + C12_eq;
                q_total = Q0_max; 
                let V_ab = q_total / C_tot;
                v_c0 = V_ab; q_c0 = C0 * V_ab;
                q_c12 = C12_eq * V_ab; v_c1 = q_c12 / C1; v_c2 = q_c12 / C2;
            }

            out_vc0.innerText = v_c0.toFixed(2);
            out_qc0.innerText = q_c0.toFixed(2);
            out_vc1.innerText = v_c1.toFixed(2);
            out_vc2.innerText = v_c2.toFixed(2);
            out_q12.innerText = q_c12.toFixed(2);
            out_qtot.innerText = q_total.toFixed(2);

            ctx.clearRect(0, 0, canvas.width, canvas.height);
            ctx.lineWidth = 2;
            ctx.strokeStyle = '#333';

            ctx.beginPath();
            // Cable Superior
            ctx.moveTo(x_bat, y_top); ctx.lineTo(x_s1_left, y_top); 
            ctx.moveTo(x_s1_right, y_top); ctx.lineTo(x_c0, y_top); 
            ctx.lineTo(x_s2_left, y_top); 
            ctx.moveTo(x_s2_right, y_top); ctx.lineTo(x_c12, y_top); 
            
            // Cable Inferior
            ctx.moveTo(x_bat, y_bot); ctx.lineTo(x_c12, y_bot); 
            
            // CONEXIONES EXACTAS A COMPONENTES
            
            // Batería
            ctx.moveTo(x_bat, y_top); ctx.lineTo(x_bat, 110); 
            ctx.moveTo(x_bat, 140); ctx.lineTo(x_bat, y_bot); 
            
            // C0 (Centro en 120, placa exterior superior en 108, placa exterior inferior en 132)
            ctx.moveTo(x_c0, y_top); ctx.lineTo(x_c0, 108); // Hasta placa superior
            ctx.moveTo(x_c0, 132); ctx.lineTo(x_c0, y_bot); // Desde placa inferior
            
            // C1 (Centro en 90, placa exterior superior en 78, placa exterior inferior en 102)
            // C2 (Centro en 160, placa exterior superior en 148, placa exterior inferior en 172)
            ctx.moveTo(x_c12, y_top); ctx.lineTo(x_c12, 78);   // Hasta placa sup C1
            ctx.moveTo(x_c12, 102); ctx.lineTo(x_c12, 148);  // Entre placa inf C1 y placa sup C2
            ctx.moveTo(x_c12, 172); ctx.lineTo(x_c12, y_bot);  // Desde placa inf C2 hasta abajo
            
            ctx.stroke();

            // Dibujar Batería
            ctx.lineWidth = 4;
            ctx.beginPath(); ctx.moveTo(x_bat - 20, 110); ctx.lineTo(x_bat + 20, 110); ctx.stroke(); 
            ctx.lineWidth = 6;
            ctx.beginPath(); ctx.moveTo(x_bat - 12, 140); ctx.lineTo(x_bat + 12, 140); ctx.stroke(); 
            ctx.fillStyle = '#000'; ctx.font = 'bold 16px Arial'; ctx.fillText(V0 + ' V', x_bat - 55, 130);

            // Interruptores
            drawSwitch(x_s1_left, x_s1_right, y_top, 'S₁', estado === 1);
            drawSwitch(x_s2_left, x_s2_right, y_top, 'S₂', estado === 2);

            // Condensadores
            drawCap(x_c0, y_c0, 'C₀', q_c0, Q0_max, '25, 135, 84');
            drawCap(x_c12, y_c1, 'C₁', q_c12, Q0_max, '13, 110, 253');
            drawCap(x_c12, y_c2, 'C₂', q_c12, Q0_max, '13, 110, 253');
        }

        slider_v0.addEventListener('input', updateSim);
        slider_c0.addEventListener('input', updateSim);
        Array.from(radios_estado).forEach(r => r.addEventListener('change', updateSim));

        updateSim();
    })();
</script>
"""

display(HTML(simulacion_analisis_final_html))
````
