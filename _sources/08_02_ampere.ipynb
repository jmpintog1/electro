
# 8.2 Ley de Ampère

Para configuraciones de corriente con alta simetría, la Ley de Ampère es una herramienta más directa que la Ley de Biot-Savart. Su forma diferencial e integral son:

$$\vec{\nabla}\times\vec{B}=\mu_{0}\vec{J}$$
$$\oint\vec{B}\cdot d\vec{l}=\mu_{0}I_{enc}$$

Donde $I_{enc}$ es la corriente neta encerrada por la curva amperiana cerrada elegida.

## Ejemplos de aplicación de la Ley de Ampère

**1. Alambre infinito**
Usando una curva amperiana circular de radio $r$ centrada en el alambre:

$$\oint B dl = B(2\pi r) = \mu_{0}I \implies B=\frac{\mu_{0}I}{2\pi r}$$

**2. Cilindro conductor de radio R**
Si una corriente constante $I$ se distribuye uniformemente en la sección transversal de un alambre grueso de radio $R$:
* **Zona interior ($r < R$):** La corriente encerrada depende del área transversal considerada.

    $$B_{i}(2\pi r) = \mu_{0}\frac{I}{\pi R^{2}}(\pi r^{2}) \implies \vec{B}_{i} = \frac{\mu_{0}I}{2\pi R^{2}}r\hat{\varphi}$$
* **Zona exterior ($r \ge R$):** Todo el alambre está encerrado.

    $$\vec{B}_{e} = \frac{\mu_{0}I}{2\pi r}\hat{\varphi}$$

**3. Solenoide Toroidal**
Para un toroide con $N$ vueltas de alambre:
* En la zona interior al agujero del toro y en la zona exterior al aparato, la corriente neta encerrada por una curva amperiana circular es cero, por lo que $B=0$.
* En el interior de las espiras del toroide, la amperiana encierra $N$ veces la corriente $I$:

    $$B(2\pi r)=\mu_{0}NI \implies \vec{B}=\frac{\mu_{0}NI}{2\pi r}\hat{\theta}$$

**4. Solenoide Recto (Bobina ideal)**
En el centro de un solenoide ideal infinitamente largo con una densidad de espiras $n = N/L$, el campo en el exterior es despreciable y en el interior es constante y uniforme:

$$B=\mu_{0}nI$$

## Fuerza entre conductores paralelos

Un alambre 1 genera un campo magnético $\vec{B}_1$ que ejerce una fuerza sobre un alambre 2 que lleva una corriente $I_2$. Si ambos son paralelos e infinitos, y están separados por una distancia $d$, la fuerza que ejerce 1 sobre una longitud $L$ de 2 es:

$$\vec{F}_{1\rightarrow 2} = \int I_{2} d\vec{l}_{2} \times \vec{B}_{1}$$
Como $\vec{B}_{1} = \frac{\mu_{0}I_{1}}{2\pi d}$, la fuerza magnética por unidad de longitud es:

$$\frac{|\vec{F}_{1\rightarrow 2}|}{L} = \frac{\mu_{0}I_{1}I_{2}}{2\pi d}$$

Por la tercera ley de Newton (acción-reacción), $\vec{F}_{1\rightarrow 2} = -\vec{F}_{2\rightarrow 1}$. Corrientes en la misma dirección se atraen; corrientes en direcciones opuestas se repelen.
