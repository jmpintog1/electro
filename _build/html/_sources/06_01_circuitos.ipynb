## Potencia Eléctrica

La energía $U$ entregada a una carga $q$ a través de un potencial $V$ está dada por $dU = dq \cdot V$. Si derivamos esto con respecto al tiempo, obtenemos la rapidez con la que se transfiere la energía, es decir, la **Potencia** ($P$):

$$P = \frac{dU}{dt} = \frac{dq}{dt} V \implies P = IV$$

La unidad de medida de la potencia es el Watt ($W$), que equivale a un Joule por segundo ($1 \, \text{W} = 1 \, \text{J/s}$).
Para el caso específico de una resistencia, usando la Ley de Ohm ($V = IR$), la potencia disipada se puede expresar como:

$$P = I^2 R$$
