
# 8.1 Ley de Biot-Savart

A diferencia de los campos eléctricos, que divergen o convergen en cargas puntuales, las líneas de campo magnético son siempre cerradas. Esto se resume en la ecuación fundamental para la divergencia del campo magnético:

$$\vec{\nabla}\cdot\vec{B}=0$$

La Ley de Biot-Savart permite calcular el campo magnético $\vec{B}$ generado por una distribución de corriente estacionaria. Para un elemento de corriente $I d\vec{l}$, la contribución $d\vec{B}$ en un punto del espacio depende del inverso del cuadrado de la distancia:

$$\vec{B}=\frac{\mu_{0}}{4\pi}\int \frac{I d\vec{l}\times\hat{r}}{r^{2}}$$

Donde $\mu_{0}$ es la permeabilidad magnética del vacío, cuyo valor es:

$$\frac{\mu_{0}}{4\pi} = 10^{-7}\text{ }\frac{\text{T}\cdot\text{m}}{\text{A}} = 10^{-7}\text{ }\frac{\text{N}}{\text{A}^{2}}$$

Una expresión más rigurosa y útil utilizando vectores de posición relativo (donde $\vec{r}$ es el punto de observación y $\vec{r} \, '$ es la posición del elemento de corriente) es:

$$\vec{B}=\frac{\mu_{0}}{4\pi}\int\frac{Id\vec{l}\times(\vec{r}-\vec{r} \, ')}{|\vec{r}-\vec{r} \, '|^{3}}$$

Si la corriente no es un hilo unidimensional, se puede generalizar usando la densidad de corriente superficial $\vec{K}$ o volumétrica $\vec{J}$.

## Ejemplos de aplicación de Biot-Savart

**1. Anillo conductor de radio R**
Para calcular el campo magnético en el eje central (eje $z$) de una espira circular de radio $R$ por la que circula una corriente $I$:

$$d\vec{l} = R d\theta (-\sin\theta \, \hat{\imath} + \cos\theta \, \hat{\jmath})$$
$$\vec{r}-\vec{r} \, ' = z\hat{k} - R(\cos\theta \, \hat{\imath} + \sin\theta \, \hat{\jmath})$$

Al integrar sobre el lazo cerrado de $0$ a $2\pi$, las componentes perpendiculares al eje se anulan por simetría, obteniendo:

$$\vec{B}(z\hat{k})=\frac{\mu_{0}I R^{2}}{2(R^{2}+z^{2})^{3/2}}\hat{k}$$

Si evaluamos el campo exactamente en el centro del anillo ($z=0$):

$$\vec{B}(\vec{0})=\frac{\mu_{0}I}{2R}\hat{k}$$

**2. Alambre recto de largo L**
Para un segmento de alambre recto a una distancia perpendicular $a$ desde su centro, integramos desde $-L/2$ hasta $L/2$:

$$\vec{B}=\frac{\mu_{0}I}{4\pi a}\frac{L}{\sqrt{a^{2}+L^{2}/4}}\hat{k}$$

Si el alambre se considera infinitamente largo ($L \rightarrow \infty$), la expresión se simplifica a:

$$\vec{B}=\frac{\mu_{0}I}{2\pi a}\hat{k}$$
