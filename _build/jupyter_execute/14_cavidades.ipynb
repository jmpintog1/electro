## Ejemplo 1: Esfera no conductora cargada con $\rho$ constante.

* **Campo eléctrico en la zona I:** $\vec{E}_{I}=\vec{0}$, porque la carga encerrada es nula.

* **Campo eléctrico en la zona II:**
    Aplicando la Ley de Gauss:
    $$\oint\vec{E}_{II}\cdot d\vec{S}=\frac{q_{enc}}{\epsilon_{0}}$$
    $$E_{II}(4\pi r^{2})=\frac{1}{\epsilon_{0}}\rho\frac{4}{3}\pi(r^{3}-a^{3})$$
    $$\vec{E}_{II}=\frac{\rho}{3\epsilon_{0}}\frac{(r^{3}-a^{3})}{r^{2}}\hat{r}$$

    Para la zona II también se puede utilizar el principio de superposición:
    $$E_{II}^{(1)}4\pi r^{2}=\frac{\rho\frac{4}{3}\pi r^{3}}{\epsilon_{0}} \implies \vec{E}_{II}^{(1)}=\frac{\rho}{3\epsilon_{0}}r\hat{r}$$
    $$E_{II}^{(2)}4\pi r^{2}=-\frac{\rho\frac{4}{3}\pi a^{3}}{\epsilon_{0}} \implies \vec{E}_{II}^{(2)}=-\frac{\rho}{3\epsilon_{0}}\frac{a^{3}}{r^{2}}\hat{r}$$
    
    Sumando ambas contribuciones obtenemos el mismo resultado:
    $$\vec{E}_{II}=\vec{E}_{II}^{(1)}+\vec{E}_{II}^{(2)}$$
    $$\vec{E}_{II}=\frac{\rho}{3\epsilon_{0}}r\hat{r}-\frac{\rho}{3\epsilon_{0}}\frac{a^{3}}{r^{2}}\hat{r}$$
    $$\vec{E}_{II}=\frac{\rho}{3\epsilon_{0}}\left(\frac{r^{3}-a^{3}}{r^{2}}\right)\hat{r}$$

* **Campo eléctrico en la zona III:**
    $$\oint E_{III}\hat{r}\cdot\hat{r}dS=\frac{q_{enc}}{\epsilon_{0}}$$
    $$E_{III}=\frac{\rho\frac{4}{3}\pi(b^{3}-a^{3})}{4\pi\epsilon_{0}r^{2}}$$
    $$\vec{E}_{III}=\frac{\rho}{3\epsilon_{0}}\frac{(b^{3}-a^{3})}{r^{2}}\hat{r}$$

## Ejemplo 2: Esfera no conductora con una cavidad esférica.

Determinar el campo eléctrico en el punto $P$, es decir, $\vec{E}_{P}$. Aplicando superposición:

1.  El campo eléctrico $\vec{E}_{1}$ para la esfera completa en el punto $P$:
    $$\oint\vec{E}_{1}\cdot d\vec{S}=\frac{q_{enc}}{\epsilon_{0}}$$
    $$E_{1}4\pi r^{2}=\frac{\rho\cdot\frac{4}{3}\pi r^{3}}{\epsilon_{0}} \implies \vec{E}_{1}=\frac{\rho}{3\epsilon_{0}}r\hat{r}$$

2.  El campo eléctrico $\vec{E}_{2}$ para el agujero en el punto $P$:
    $$E_{2}4\pi r^{\prime 2}=\frac{\rho\frac{4}{3}\pi r^{\prime 3}}{\epsilon_{0}} \implies \vec{E}_{2}=\frac{\rho}{3\epsilon_{0}}r^{\prime}\hat{r}^{\prime}$$

Dado que la relación vectorial de las distancias es $\vec{d}+\vec{r}^{\prime}=\vec{r}$, el campo resultante es:
$$\vec{E}_{P}=\vec{E}_{1}-\vec{E}_{2}$$
$$\vec{E}_{P}=\frac{\rho}{3\epsilon_{0}}\vec{r}-\frac{\rho}{3\epsilon_{0}}\vec{r}^{\prime}$$
$$\vec{E}_{P}=\frac{\rho}{3\epsilon_{0}}(\vec{r}-\vec{r}^{\prime})$$
$$\vec{E}_{P}=\frac{\rho}{3\epsilon_{0}}\vec{d}$$

**Análisis Conceptual:**
La pregunta plantea un error conceptual común al aplicar la Ley de Gauss. A pesar de que la carga encerrada en el agujero es cero, el campo eléctrico en su interior no lo es. Esto se debe a que el campo eléctrico en un punto no depende solo de la carga encerrada en una superficie gaussiana que lo contenga, sino de todas las cargas de la distribución, tanto dentro como fuera de la superficie.

La Ley de Gauss establece que el flujo eléctrico neto a través de una superficie cerrada es proporcional a la carga neta encerrada: $\Phi_{E}=\oint\vec{E}\cdot d\vec{S}=q_{enc}/\epsilon_{0}$. Sin embargo, esta ley no implica que el campo sea cero si $q_{enc}=0$. Para que el campo eléctrico sea cero en una región sin carga, se necesita una simetría esférica en la distribución de carga que rodea esa región. Un ejemplo perfecto es el interior de una esfera conductora o una cáscara esférica aislante con carga uniforme. En estos casos, para cualquier superficie gaussiana dentro, las líneas de campo que entran compensan exactamente a las que salen, dando un flujo total nulo y, por lo tanto, un campo eléctrico cero.

En nuestro problema, la carga no está distribuida simétricamente alrededor del agujero. El agujero está desplazado con respecto al centro de la esfera grande. Esta asimetría es la clave. La carga fuera del agujero (el material aislante con densidad de carga $\rho$) contribuye al campo eléctrico en el interior del agujero, lo que hace que el campo sea no nulo y uniforme, como se demostró en el problema anterior usando el principio de superposición.
