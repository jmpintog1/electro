# 4.3 Energía en un Condensador

El proceso de cargar un condensador implica realizar trabajo para trasladar carga de un conductor a otro en contra del campo eléctrico existente. El trabajo infinitesimal $dW$ requerido para transferir una carga $dq$ cuando la diferencia de potencial es $V = q/C$ está dado por:

$$dW = V dq = \frac{q}{C} dq$$

El trabajo total para cargar el condensador desde $q=0$ hasta $q=Q$ es:

$$W = \int_{0}^{Q} \frac{q}{C} dq = \frac{1}{2}\frac{Q^2}{C}$$

Este trabajo se almacena como energía potencial eléctrica $U$:

$$U = \frac{1}{2}\frac{Q^2}{C} = \frac{1}{2}QV = \frac{1}{2}CV^2$$

## Densidad de energía
Si analizamos el condensador de placas paralelas, donde $U = \frac{1}{2} C (\Delta V)^2 = \frac{1}{2} \left( \frac{\epsilon_0 A}{d} \right) (Ed)^2 = \frac{1}{2} \epsilon_0 (Ad) E^2$, podemos definir la **densidad de energía** $u$ (energía por unidad de volumen). Como el volumen es $\mathcal{V} = Ad$:

$$u = \frac{dU}{d\mathcal{V}} = \frac{1}{2}\epsilon_0 E^2$$

Este resultado, aunque derivado para placas paralelas, es válido para cualquier campo eléctrico en el vacío.
