# 3.3 Energía Potencial y Trabajo

El potencial $V$ es una magnitud que se comprende mejor al vincularla con la energía potencial electrostática $U$. Dimensionalmente, el potencial se define como la energía por unidad de carga:

$$dim(V)=\frac{dim(\text{energía})}{dim(\text{carga})}$$

De esta forma, si ubicamos una carga $q$ en una región con potencial $V$, la energía potencial de esa carga es:

$$U = q V$$

Sabiendo que la fuerza electrostática es conservativa, el trabajo $W_q$ que esta realiza sobre una partícula equivale al negativo de la variación de su energía potencial:

$$W_q=-\Delta U_q=-(U_f - U_i)=-q \, \Delta V$$

> **Unidades:** En el Sistema Internacional, la diferencia de potencial se mide en Volts (V), equivalente a 1 J/C. Sin embargo, a escala atómica es muy frecuente utilizar el electrón-voltio (eV), donde $1 \, \text{eV} = 1{,}6 \times 10^{-19} \, \text{C} \cdot 1 \, \text{V}$.

## Trabajo de un agente externo

### Velocidad constante:

Imaginemos el proceso de mover una carga de prueba $q_0$ desde un punto muy lejano ($i \rightarrow \infty$, donde asumimos que $V_i = 0$) hasta un punto $f$. El teorema del trabajo y la energía cinética nos dice que $W_{\text{Total}} = \Delta K$. 

Si un agente externo mueve la carga con velocidad constante, el cambio de energía cinética es nulo ($\Delta K = 0$), lo que implica que el trabajo total sobre el sistema es cero.

$$\begin{align}
W_q + W_{\text{ext}} &= 0 \\
-\Delta U_q + W_{\text{ext}} &= 0
\end{align}$$

Por consiguiente, el trabajo que debe realizar el agente externo es exactamente:

$$W_{\text{ext}} = q \, \Delta V$$

### Campo eléctrico constante:

Si movemos una carga $q$ a través de una distancia $d$ en un campo eléctrico constante $\vec{E}$, con una aceleración constante, esta aumentará su energía cinética y disminuirá su energía potencial eléctrica:

$$\Delta V=-Ed$$

$$\Delta U_q=-qEd$$


## Energía Electrostática

### Sistema de Cargas Discretas

Para un sistema discreto, si analizamos el trabajo realizado para traer las cargas desde el infinito y armar el sistema, la energía para dos cargas $q_1$ y $q_2$ es:

$$U = \frac{1}{4\pi\epsilon_0}\frac{q_1 q_2}{|\vec{r}_2 - \vec{r}_1|} = q_1 V_2 = q_2 V_1$$

El signo de la energía potencial resultante tiene un significado físico directo:
* Si $U > 0$, las fuerzas entre las cargas son de repulsión.
* Si $U < 0$, las fuerzas son de atracción.

Para incorporar una tercera carga $q_3$ al sistema, el trabajo requerido es $W_3 = q_3(V_1 + V_2)$. Así, la energía total de un sistema de tres cargas discretas queda conformada por la superposición de los pares:

$$U = \frac{1}{4\pi\epsilon_0} \left( \frac{q_1 q_2}{|\vec{r}_2 - \vec{r}_1|} + \frac{q_2 q_3}{|\vec{r}_3 - \vec{r}_1|} + \frac{q_2 q_3}{|\vec{r}_3 - \vec{r}_2|} \right)$$

Escribiendo esto de manera general para cualquier cantidad de cargas, la energía electrostática se expresa como:

$$U = \frac{1}{2} \frac{1}{4\pi\epsilon_0} \sum_{i \neq j} \frac{q_i q_j}{|\vec{r}_j - \vec{r}_i|} = \frac{1}{2} \sum_{i \neq j} q_i V_j$$

El término $\frac{1}{2}$ se introduce debido a que en la sumatoria se considera cada par de cargas dos veces.

### Distribución Continua de Carga

Considerando el caso continuo, la sumatoria se transforma en una integral sobre toda la distribución:

$$U = \frac{1}{2} \int V dq$$

#### Energía de una esfera uniformemente cargada

Consideremos una esfera de radio $R$ con una carga total $Q$ distribuida uniformemente en su volumen. Iremos armando esta esfera construyendo capas sucesivas, de modo que el potencial irá aumentando progresivamente.
El trabajo diferencial es $dW = V dq$, por lo que la energía total equivale a $W = \int V dq$.

Utilizando la densidad volumétrica $\rho = \frac{Q}{\frac{4}{3}\pi R^3}$, el diferencial de carga es $dq = \rho d\nu$.
El potencial en la superficie de un radio intermedio $r$ generado por la carga encerrada es:

$$V = \frac{\rho \frac{4}{3}\pi r^3}{4\pi\epsilon_0 r}$$

Integrando el trabajo a través de todo el volumen (desde $0$ hasta $R$):

$$\begin{aligned} W &= \int_0^R \frac{\rho \frac{4}{3}\pi r^3}{4\pi\epsilon_0 r} (\rho 4\pi r^2) dr \\
W &= \frac{\rho^2 \frac{4\pi}{3} 4\pi}{4\pi\epsilon_0} \int_0^R r^4 dr \\
W &= \frac{\rho^2 \frac{4\pi}{3}}{\epsilon_0} \frac{R^5}{5}\end{aligned}$$

Reemplazando $\rho$ nuevamente en términos de la carga total $Q$, encontramos que la energía almacenada en esta distribución de carga es:

$$U = W = \frac{3}{5} \frac{Q^2}{4\pi\epsilon_0 R}$$

*Método Alternativo:* Podemos llegar al mismo resultado utilizando la ecuación general $U = \frac{1}{2} \int \rho d\nu \cdot V$ integrando sobre todo el volumen con el potencial interno de la esfera $V_I = \frac{Q}{8\pi\epsilon_0 R^3}(3R^2 - r^2)$:

$$\begin{aligned} U &= \frac{1}{2} \int_0^R (\rho 4\pi r^2 dr) \frac{Q}{8\pi\epsilon_0 R^3}(3R^2 - r^2) \\
U &= \frac{1}{2} \rho \frac{4\pi Q}{8\pi\epsilon_0 R^3} \frac{4}{5}R^5 = \frac{3Q^2}{5 \cdot 4\pi\epsilon_0 R}\end{aligned}$$

Se verifica analíticamente que se obtiene el mismo resultado anterior.
