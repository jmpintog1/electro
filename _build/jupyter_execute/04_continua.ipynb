# Distribución de Carga


En casos donde la carga no es puntual sino que se distribuye en un volumen, superficie o línea, el campo eléctrico en el punto $P$ se calcula mediante una integración sobre toda la distribución:

$$
\begin{aligned}
	d\vec{E}_P &= \frac{1}{4\pi\epsilon_0} \frac{dq'}{|\vec{r}-\vec{r}\,'|^2} \frac{(\vec{r}-\vec{r}\,')}{|\vec{r}-\vec{r}\,'|} \\
	\vec{E}_P &= \frac{1}{4\pi\epsilon_0} \int \frac{dq'(\vec{r}-\vec{r}\,')}{|\vec{r}-\vec{r}\,'|^3} \\
	\vec{E}_P &= \frac{1}{4\pi\epsilon_0} \int dq' \frac{(\vec{r}-\vec{r}\,')}{|\vec{r}-\vec{r}\,'|^3}
\end{aligned}
$$

### Guía rápida: ¿Cuál densidad utilizar?

En la resolución de problemas de ingeniería, identificar la geometría es el primer paso para plantear la integral de campo eléctrico.

| Dimensión | Tipo de Densidad | Símbolo | Geometrías Comunes | Diferencial de carga ($dq'$) |
| :--- | :--- | :---: | :--- | :--- |
| **1D** | Lineal | $\lambda$ | Hilos, anillos, varillas delgadas. | $dq' = \lambda dl$ |
| **2D** | Superficial | $\sigma$ | Discos, láminas, cáscaras esféricas. | $dq' = \sigma ds$ |
| **3D** | Volumétrica | $\rho$ | Esferas sólidas, cilindros macizos. | $dq' = \rho dV$ |

---

> **💡 Consejo para el cálculo:** Recuerda que si la carga está distribuida de manera **uniforme**, puedes calcular la densidad como la carga total dividida por la longitud, área o volumen total ($Q/L$, $Q/A$ o $Q/V$).

````python
import numpy as np
import matplotlib.pyplot as plt
import ipywidgets as widgets
from ipywidgets import FloatSlider, IntSlider, Output, VBox
from matplotlib.patches import Rectangle

# Configuración inicial de constantes (simplificadas para visualización visual)
k = 8.99e9  # N m^2 / C^2
lam = 1e-6 # Densidad lineal uniforme (1 uC/m)
bar_L = 2.0 # Longitud de la barra

def simular_superposicion_campo(px, py, num_seg, seg_actual):
    plt.close('all') # Limpiar figuras anteriores para fluidez
    
    # 1. Discretizar la barra en segmentos diferenciales finitos
    dx = bar_L / num_seg
    # Coordenadas x' de los centros de cada segmento
    fuentes_x = np.linspace(dx/2, bar_L - dx/2, num_seg)
    
    # 2. Calcular las contribuciones individuales y acumuladas
    contribuciones_E = []
    
    # Singularity check: si P está sobre la barra
    dist_minima_seg = np.sqrt((px - fuentes_x)**2 + py**2)
    if np.any(dist_minima_seg < dx*0.9):
        print("⚠️ El punto P está demasiado cerca de un elemento de la barra. División por cero evitada.")
        return

    for x_prime in fuentes_x:
        # Vector r_P - r_fuente
        dx_vec = px - x_prime
        dy_vec = py - 0 # Barra en eje X, y'=0
        
        # Distancia d^2 y d^3
        d2 = dx_vec**2 + dy_vec**2
        d3 = d2**(1.5)
        
        # dE = k * lambda * dx * r_vec / d^3
        dEx = k * lam * dx * dx_vec / d3
        dEy = k * lam * dx * dy_vec / d3
        contribuciones_E.append(np.array([dEx, dEy]))

    # Convertir a array para manejo fácil
    contribuciones_E = np.array(contribuciones_E)
    
    # Campo total final (teórico sumando todos)
    E_total_final = np.sum(contribuciones_E, axis=0)
    
    # Campo acumulado hasta el segmento seleccionado (la integral parcial)
    if seg_actual >= num_seg: seg_actual = num_seg - 1
    E_acumulado = np.sum(contribuciones_E[:seg_actual+1], axis=0)
    
    # dE específico del segmento actual (flecha roja)
    dE_actual = contribuciones_E[seg_actual]

    # --- GRAFICACIÓN ---
    fig, ax = plt.subplots(figsize=(10, 8))
    
    # A) Dibujar la Barra
    # Barra base (gris)
    ax.add_patch(Rectangle((0, -0.05), bar_L, 0.1, color='lightgray', zorder=1, label='Barra cargada'))
    
    # Resaltar todos los segmentos analizados hasta ahora (verde suave)
    resaltado_w = fuentes_x[seg_actual] + dx/2
    ax.add_patch(Rectangle((0, -0.05), resaltado_w, 0.1, color='green', alpha=0.3, zorder=2))
    
    # Resaltar el segmento infinitesimal dx' actual (rojo)
    x_rect = fuentes_x[seg_actual] - dx/2
    ax.add_patch(Rectangle((x_rect, -0.05), dx, 0.1, color='red', zorder=3, label="Elemento dx' actual"))

    # B) Dibujar Punto P
    ax.scatter([px], [py], color='black', s=100, zorder=10)
    ax.text(px + 0.1, py + 0.1, 'P', fontsize=14, fontweight='bold')
    
    # C) Dibujar Vectores de Campo (Normalizados para visualización)
    scale_factor = 0.8 / max(np.linalg.norm(E_total_final), np.linalg.norm(dE_actual), 1e-9)
    
    # 1. dE actual (rojo) - Usando doble comilla
    ax.quiver(px, py, dE_actual[0]*scale_factor, dE_actual[1]*scale_factor, 
              angles='xy', scale_units='xy', scale=1, color='red', width=0.012, 
              label=r"d$\vec{E}$ actual")
    
    # 2. E acumulado (verde) - SOLUCIONADO: Usando comillas dobles r"..."
    ax.quiver(px, py, E_acumulado[0]*scale_factor, E_acumulado[1]*scale_factor, 
              angles='xy', scale_units='xy', scale=1, color='green', alpha=0.6, width=0.015,
              label=r"$\vec{E}_{acumulada}$ ($\int_0^{x'} d\vec{E}$)")
    
    # 3. E total final (negro punteado) - Usando doble comilla
    ax.quiver(px, py, E_total_final[0]*scale_factor, E_total_final[1]*scale_factor, 
              angles='xy', scale_units='xy', scale=1, color='black', linestyle=':', 
              linewidth=2, width=0.008, alpha=0.4, label=r"$\vec{E}_{total\_final}$")
    
    # D) Detalles del gráfico
    mag_E_final = np.linalg.norm(E_total_final)
    ax.set_title(f'Superposición de Campo Eléctrico en P({px:.1f}, {py:.1f})\n' +
                 f'Intensidad Total: {mag_E_final:.2e} N/C', fontsize=12)
    ax.set_xlim(-1, bar_L + 1)
    ax.set_ylim(-3, 3) # Eje Y amplio para colocar P arriba/abajo
    ax.set_xlabel('x (m)')
    ax.set_ylabel('y (m)')
    ax.grid(True, alpha=0.2)
    ax.set_aspect('equal') # Importante para que los vectores se vean geométricamente correctos
    ax.legend(loc='upper right', fontsize=10)
    plt.show()

# --- Configuración de Controles ---
out = Output()

def update_sim(px_val, py_val, n_segments, curr_idx):
    with out:
        out.clear_output(wait=True)
        simular_superposicion_campo(px_val, py_val, n_segments, curr_idx)

sl_px = FloatSlider(min=-0.8, max=bar_L+0.8, step=0.1, value=1.0, description='Posición Px:')
sl_py = FloatSlider(min=-2.5, max=2.5, step=0.1, value=1.5, description='Posición Py:')
sl_n = IntSlider(min=10, max=100, step=10, value=50, description='Nº Segmentos:')
sl_idx = IntSlider(min=0, max=49, step=1, value=0, description='Elemento actual:')

# Vincular el slider de índice al número de segmentos para que el rango coincida
def update_idx_range(*args):
    sl_idx.max = sl_n.value - 1
sl_n.observe(update_idx_range, 'value')

# Usar interactive_output para una disposición más limpia
interact_plot = widgets.interactive_output(update_sim, 
                                           {'px_val': sl_px, 'py_val': sl_py, 
                                            'n_segments': sl_n, 'curr_idx': sl_idx})

display(VBox([interact_plot, sl_px, sl_py, sl_n, sl_idx]))
````
