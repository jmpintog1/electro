# Distribución de Carga


En casos donde la carga no es puntual sino que se distribuye en un volumen, superficie o línea, el campo eléctrico en el punto $P$ se calcula mediante una integración sobre toda la distribución:

$$
\begin{aligned}
	d\vec{E}_P &= \frac{1}{4\pi\epsilon_0} \frac{dq'}{|\vec{r}-\vec{r}\,'|^2} \frac{(\vec{r}-\vec{r}\,')}{|\vec{r}-\vec{r}\,'|} \\
	\vec{E}_P &= \frac{1}{4\pi\epsilon_0} \int \frac{dq'(\vec{r}-\vec{r}\,')}{|\vec{r}-\vec{r}\,'|^3} \\
	\vec{E}_P &= \frac{1}{4\pi\epsilon_0} \int dq' \frac{(\vec{r}-\vec{r}\,')}{|\vec{r}-\vec{r}\,'|^3}
\end{aligned}
$$

### Guía rápida: ¿Cuál densidad utilizar?

En la resolución de problemas de ingeniería, identificar la geometría es el primer paso para plantear la integral de campo eléctrico.

| Dimensión | Tipo de Densidad | Símbolo | Geometrías Comunes | Diferencial de carga ($dq'$) |
| :--- | :--- | :---: | :--- | :--- |
| **1D** | Lineal | $\lambda$ | Hilos, anillos, varillas delgadas. | $dq' = \lambda dl$ |
| **2D** | Superficial | $\sigma$ | Discos, láminas, cáscaras esféricas. | $dq' = \sigma ds$ |
| **3D** | Volumétrica | $\rho$ | Esferas sólidas, cilindros macizos. | $dq' = \rho dV$ |

---

> **💡 Consejo para el cálculo:** Recuerda que si la carga está distribuida de manera **uniforme**, puedes calcular la densidad como la carga total dividida por la longitud, área o volumen total ($Q/L$, $Q/A$ o $Q/V$).

````python
import numpy as np
import matplotlib.pyplot as plt
import ipywidgets as widgets
from ipywidgets import FloatSlider, IntSlider, Checkbox, VBox, HBox
from matplotlib.patches import Rectangle

# Configuración inicial de constantes
k = 8.99e9  # N m^2 / C^2
lam = 1e-6 # Densidad lineal uniforme (1 uC/m)
bar_L = 2.0 # Longitud de la barra

def simular_superposicion_campo(px, py, num_seg, seg_actual, show_dE, show_Eac, show_Etot):
    plt.close('all') 
    
    dx = bar_L / num_seg
    fuentes_x = np.linspace(dx/2, bar_L - dx/2, num_seg)
    
    contribuciones_E = []
    
    # Singularity check 
    dist_minima_seg = np.sqrt((px - fuentes_x)**2 + py**2)
    if np.any(dist_minima_seg < dx*0.9):
        print("⚠️ El punto P está demasiado cerca de un elemento de la barra. Aleja P.")
        return

    for x_prime in fuentes_x:
        dx_vec = px - x_prime
        dy_vec = py - 0 
        
        d2 = dx_vec**2 + dy_vec**2
        d3 = d2**(1.5)
        
        dEx = k * lam * dx * dx_vec / d3
        dEy = k * lam * dx * dy_vec / d3
        contribuciones_E.append(np.array([dEx, dEy]))

    contribuciones_E = np.array(contribuciones_E)
    E_total_final = np.sum(contribuciones_E, axis=0)
    
    if seg_actual >= num_seg: seg_actual = num_seg - 1
    E_acumulado = np.sum(contribuciones_E[:seg_actual+1], axis=0)
    dE_actual = contribuciones_E[seg_actual]

    # --- GRAFICACIÓN ---
    fig, ax = plt.subplots(figsize=(12, 10)) 
    
    # Dibujar la Barra
    ax.add_patch(Rectangle((0, -0.05), bar_L, 0.1, color='lightgray', zorder=1))
    resaltado_w = fuentes_x[seg_actual] + dx/2
    ax.add_patch(Rectangle((0, -0.05), resaltado_w, 0.1, color='green', alpha=0.3, zorder=2))
    x_rect = fuentes_x[seg_actual] - dx/2
    ax.add_patch(Rectangle((x_rect, -0.05), dx, 0.1, color='red', zorder=3))

    # Dibujar Punto P
    ax.scatter([px], [py], color='black', s=100, zorder=10)
    ax.text(px + 0.2, py + 0.2, 'P', fontsize=14, fontweight='bold')
    
    # Factor de escala base
    dist_ref = max(np.sqrt((px - bar_L/2)**2 + py**2), 1.0)
    scale_factor = (dist_ref * 1.2) / max(np.linalg.norm(E_total_final), 1e-9)
    
    # MAGNIFICACIÓN DINÁMICA DEL VECTOR ROJO
    # Ampliamos visualmente el dE para que su dirección sea siempre visible
    mag_red = max(1.0, num_seg / 4.0) 

    # Dibujar vectores condicionalmente
    if show_dE:
        ax.quiver(px, py, dE_actual[0]*scale_factor*mag_red, dE_actual[1]*scale_factor*mag_red, 
                  angles='xy', scale_units='xy', scale=1, color='red', 
                  width=0.009, headwidth=5, headlength=6, zorder=6,
                  label=f"d$\\vec{{E}}$ actual (Magnificado x{mag_red:.0f})")
        
    if show_Eac:
        ax.quiver(px, py, E_acumulado[0]*scale_factor, E_acumulado[1]*scale_factor, 
                  angles='xy', scale_units='xy', scale=1, color='green', alpha=0.6, 
                  width=0.015, headwidth=4, headlength=5, zorder=5,
                  label=r"$\vec{E}_{acumulada}$")
        
    if show_Etot:
        ax.quiver(px, py, E_total_final[0]*scale_factor, E_total_final[1]*scale_factor, 
                  angles='xy', scale_units='xy', scale=1, color='black', linestyle=':', 
                  linewidth=2.5, width=0.008, headwidth=4, headlength=5, alpha=0.5, zorder=4,
                  label=r"$\vec{E}_{total\_final}$")
    
    mag_E_final = np.linalg.norm(E_total_final)
    ax.set_title(f'Superposición de Campo Eléctrico en P({px:.1f}, {py:.1f})\n' +
                 f'Intensidad Total: {mag_E_final:.2e} N/C', fontsize=14)
    
    x_min = min(-2, px - 3)
    x_max = max(bar_L + 2, px + 3)
    y_min = min(-4, py - 3)
    y_max = max(4, py + 3)
    
    ax.set_xlim(x_min, x_max)
    ax.set_ylim(y_min, y_max) 
    ax.set_xlabel('x (m)')
    ax.set_ylabel('y (m)')
    ax.grid(True, alpha=0.3)
    ax.set_aspect('equal') 
    
    if show_dE or show_Eac or show_Etot:
        ax.legend(loc='best', fontsize=12)
    
    plt.show()

# --- Configuración de Controles ---
sl_px = FloatSlider(min=-5.0, max=7.0, step=0.1, value=1.0, description='Posición Px:')
sl_py = FloatSlider(min=-5.0, max=5.0, step=0.1, value=2.0, description='Posición Py:')
# Cambiado por defecto a 20 para que los saltos sean más notorios
sl_n = IntSlider(min=10, max=100, step=10, value=20, description='Nº Segmentos:') 
sl_idx = IntSlider(min=0, max=19, step=1, value=0, description='Elemento dx\':')

chk_dE = Checkbox(value=True, description='Mostrar dE actual (Rojo)')
chk_Eac = Checkbox(value=True, description='Mostrar E acumulado (Verde)')
chk_Etot = Checkbox(value=True, description='Mostrar E total final (Negro)')

def update_idx_range(*args):
    sl_idx.max = sl_n.value - 1
    if sl_idx.value > sl_n.value - 1:
        sl_idx.value = sl_n.value - 1
sl_n.observe(update_idx_range, 'value')

interact_plot = widgets.interactive_output(simular_superposicion_campo, 
                                           {'px': sl_px, 'py': sl_py, 
                                            'num_seg': sl_n, 'seg_actual': sl_idx,
                                            'show_dE': chk_dE, 'show_Eac': chk_Eac, 'show_Etot': chk_Etot})

controles_posicion = HBox([sl_px, sl_py])
controles_segmentos = HBox([sl_n, sl_idx])
controles_visibilidad = HBox([chk_dE, chk_Eac, chk_Etot])

display(VBox([interact_plot, controles_posicion, controles_segmentos, controles_visibilidad]))
````
