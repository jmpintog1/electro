# 7.1 Magnetismo

## Introducción al Magnetismo

Históricamente, la fuente natural de campo magnético es la "magnetita". Los materiales como el hierro también están fuertemente asociados a estos fenómenos. Todo imán tiene dos polos, conocidos como Norte (N) y Sur (S). Las propiedades fundamentales de estos polos establecen que los polos iguales se repelen y los polos opuestos se atraen. Además, los polos magnéticos no se pueden separar.

## Fuerza Magnética y Campo Magnético

Cuando una carga puntual $q$ se mueve con una velocidad $\vec{v}$ en presencia de un campo magnético $\vec{B}$, experimenta la fuerza de Lorentz (en su componente estrictamente magnética), descrita por la ecuación:

$$\vec{F}_{\text{L}} = q\, \vec{v} \times \vec{B}$$

De este producto cruz se deduce que:
* Si la velocidad es paralela al campo magnético ($\vec{v} \parallel \vec{B}$), la fuerza resultante es nula ($\vec{F}_{\text{L}} = \vec{0}$).
* La magnitud de la fuerza está dada por $F_{\text{L}} = |q|vB\sin(\theta)$, donde $\theta$ es el ángulo entre los vectores $\vec{v}$ y $\vec{B}$.

**Análisis Dimensional y Unidades**

El análisis dimensional para el campo magnético $\vec{B}$ resulta en:

$$dim(B) = \frac{\text{fuerza}}{\text{carga} \cdot \left(\frac{\text{longitud}}{\text{tiempo}}\right)} = \frac{\text{fuerza} \cdot \text{tiempo}}{\text{carga} \cdot \text{longitud}} = \frac{\text{fuerza}}{\text{corriente} \cdot \text{longitud}}$$

En el Sistema Internacional (SI), la unidad de medida para el campo magnético es el Tesla (T).

$$1\text{ T} = 1\frac{\text{N}}{\text{A} \cdot \text{m}}$$

Como referencia, el campo magnético terrestre tiene una magnitud aproximada de $B \approx 10^{-4}\text{ T}$.

## Movimiento en un Campo Magnético Uniforme

Cuando una carga ingresa a una zona con un campo magnético constante, la fuerza magnética perpendicular desvía su trayectoria, iniciando un movimiento circular. La fuerza de Lorentz actúa como fuerza centrípeta, por lo que se cumple:

$$F_{\text{L}} = qvB = ma_{c} = \frac{mv^{2}}{R}$$

A partir de esto, podemos determinar que el radio $R$ de la trayectoria es:

$$R = \frac{mv}{qB}$$

Dado que la velocidad angular está relacionada por $\omega = \frac{v}{R}$, el periodo orbital $T$ de la carga está dado por:

$$T = \frac{2\pi m}{qB}$$

### Ejemplo

Para un electrón sometido a un campo de $B = 10^{-4}\text{ T}$:
* Masa: $m_{e} = 9{,}11 \times 10^{-31}\text{ kg}$
* Carga: $q = 1{,}6 \times 10^{-19}\text{ C}$

El cálculo estimado del periodo resulta en:

$$T = \frac{2\pi \cdot 9{,}11 \times 10^{-31}}{1{,}6 \times 10^{-19} \cdot 10^{-4}} \approx 10^{-7}\text{ s} \approx 10\text{ }\mu\text{s}$$

```python
import random
from IPython.display import display, HTML

# UID para evitar colisiones en Jupyter Book
uid = str(random.randint(10000, 99999))

simulacion_html = """
<div style="border: 1px solid #e0e0e0; padding: 20px; border-radius: 8px; background-color: #f8f9fa; font-family: -apple-system, sans-serif;">
    <h3 style="margin-top:0; color: #2c3e50;">Simulación: Partícula en Campo Magnético Constante</h3>
    <p style="font-size: 0.95em; color: #555; margin-bottom: 15px;">Ajusta los parámetros y presiona <b>"Lanzar Partícula"</b>.</p>
    
    <div style="display: flex; gap: 20px; flex-wrap: wrap; margin-bottom: 15px;">
        <div style="flex: 1.5; min-width: 300px;">
            <div style="margin-bottom: 10px;">
                <label style="font-weight: 600; display: inline-block; width: 140px;">Carga (q):</label>
                <select id="charge_UID" style="padding: 4px; width: 45%;">
                    <option value="1">Positiva (+)</option>
                    <option value="-1">Negativa (-)</option>
                    <option value="0">Neutra (0)</option>
                </select>
            </div>
            <div style="margin-bottom: 8px;">
                <label style="font-weight: 600; display: inline-block; width: 140px;">Campo (B): <span id="valB_UID" style="font-weight: normal;">2.0</span></label>
                <input type="range" id="B_UID" min="-5" max="5" step="0.5" value="2.0" style="width: 45%; vertical-align: middle;">
            </div>
            <div style="margin-bottom: 8px;">
                <label style="font-weight: 600; display: inline-block; width: 140px;">Masa (m): <span id="valM_UID" style="font-weight: normal;">2.0</span></label>
                <input type="range" id="M_UID" min="1" max="10" step="0.5" value="2.0" style="width: 45%; vertical-align: middle;">
            </div>
            <div style="margin-bottom: 8px;">
                <label style="font-weight: 600; display: inline-block; width: 140px;">Velocidad (v₀): <span id="valV_UID" style="font-weight: normal;">3.0</span></label>
                <input type="range" id="V_UID" min="1" max="6" step="0.5" value="3.0" style="width: 45%; vertical-align: middle;">
            </div>
        </div>

        <div style="flex: 1; min-width: 220px; display: flex; flex-direction: column; justify-content: center; gap: 10px;">
            <button id="launch_UID" style="background-color: #f73bc5; color: white; border: none; padding: 10px 15px; border-radius: 5px; cursor: pointer; font-weight: bold;">🚀 Lanzar Partícula</button>
            <button id="clear_UID" style="background-color: #6c757d; color: white; border: none; padding: 8px 15px; border-radius: 5px; cursor: pointer;">Limpiar Trayectorias</button>
        </div>
    </div>
    
    <div style="text-align: center; position: relative; overflow: hidden;">
        <canvas id="canvas_UID" width="800" height="500" style="background: #ffffff; border: 1px solid #ccc; border-radius: 4px; box-shadow: 0 2px 5px rgba(0,0,0,0.05); max-width: 100%;"></canvas>
    </div>
</div>

<script>
    (function() {
        function initSimulation() {
            const canvas = document.getElementById('canvas_UID');
            if (!canvas) {
                setTimeout(initSimulation, 50); 
                return;
            }
            
            const ctx = canvas.getContext('2d');
            const in_charge = document.getElementById('charge_UID');
            const in_B = document.getElementById('B_UID');
            const in_M = document.getElementById('M_UID');
            const in_V = document.getElementById('V_UID');
            
            const val_B = document.getElementById('valB_UID');
            const val_M = document.getElementById('valM_UID');
            const val_V = document.getElementById('valV_UID');
            
            const btn_launch = document.getElementById('launch_UID');
            const btn_clear = document.getElementById('clear_UID');
            
            let particles = [];
            let deadPaths = []; 
            let animationId;
            
            function updateLabels() {
                val_B.innerText = parseFloat(in_B.value).toFixed(1);
                val_M.innerText = parseFloat(in_M.value).toFixed(1);
                val_V.innerText = parseFloat(in_V.value).toFixed(1);
            }
            
            in_charge.addEventListener('change', updateLabels);
            in_B.addEventListener('input', updateLabels);
            in_M.addEventListener('input', updateLabels);
            in_V.addEventListener('input', updateLabels);
            
            function drawMagneticField() {
                let B = in_B ? parseFloat(in_B.value) : 2.0;
                if (B === 0) return;
                
                ctx.fillStyle = '#a0aabc';
                ctx.font = '14px Arial';
                ctx.textAlign = 'center';
                ctx.textBaseline = 'middle';
                const spacing = 40;
                
                // CORRECCIÓN: B > 0 representa campo hacia afuera (círculos), B < 0 campo hacia adentro (cruces)
                const symbol = B > 0 ? '·' : '×';
                
                for (let x = spacing/2; x < canvas.width; x += spacing) {
                    for (let y = spacing/2; y < canvas.height; y += spacing) {
                        if (B > 0) { // Ahora dibujamos el círculo cuando B es positivo (hacia afuera)
                            ctx.beginPath();
                            ctx.arc(x, y, 4, 0, Math.PI*2);
                            ctx.strokeStyle = '#a0aabc';
                            ctx.stroke();
                        }
                        ctx.fillText(symbol, x, y);
                    }
                }
            }
            
            btn_launch.addEventListener('click', () => {
                let q = parseInt(in_charge.value);
                let color = q > 0 ? '#dc3545' : (q < 0 ? '#0d6efd' : '#212529');
                let initialSpeed = parseFloat(in_V.value);
                
                particles.push({
                    x: 0, y: canvas.height / 2,
                    vx: initialSpeed, vy: 0,
                    initial_v: initialSpeed, 
                    q: q, m: parseFloat(in_M.value),
                    color: color,
                    path: [{x: 0, y: canvas.height / 2}]
                });
                if (!animationId) animate();
            });
            
            btn_clear.addEventListener('click', () => {
                particles = [];
                deadPaths = []; 
                ctx.clearRect(0, 0, canvas.width, canvas.height);
                drawMagneticField();
            });
            
            function animate() {
                ctx.clearRect(0, 0, canvas.width, canvas.height);
                drawMagneticField();
                
                deadPaths.forEach(dp => {
                    if (dp.path.length < 2) return;
                    ctx.beginPath();
                    ctx.moveTo(dp.path[0].x, dp.path[0].y);
                    for (let j = 1; j < dp.path.length; j++) {
                        ctx.lineTo(dp.path[j].x, dp.path[j].y);
                    }
                    ctx.strokeStyle = dp.color;
                    ctx.lineWidth = 2;
                    ctx.setLineDash([5, 5]);
                    ctx.stroke();
                    ctx.setLineDash([]);
                });
                
                let B_current = parseFloat(in_B.value);
                let scale_factor = 0.03; 
                
                for (let i = particles.length - 1; i >= 0; i--) {
                    let p = particles[i];
                    
                    let ax = (p.q / p.m) * (-p.vy * B_current) * scale_factor;
                    let ay = (p.q / p.m) * (p.vx * B_current) * scale_factor;
                    
                    p.vx += ax;
                    p.vy += ay;
                    
                    let current_speed = Math.sqrt(p.vx * p.vx + p.vy * p.vy);
                    if (current_speed > 0 && p.initial_v > 0) {
                        p.vx = (p.vx / current_speed) * p.initial_v;
                        p.vy = (p.vy / current_speed) * p.initial_v;
                    }
                    
                    p.x += p.vx;
                    p.y += p.vy;
                    
                    if (p.path.length === 0 || 
                       Math.abs(p.x - p.path[p.path.length-1].x) > 2 || 
                       Math.abs(p.y - p.path[p.path.length-1].y) > 2) {
                        p.path.push({x: p.x, y: p.y});
                    }
                    
                    if (p.x < 0 || p.x > canvas.width || p.y < 0 || p.y > canvas.height) {
                        deadPaths.push({path: p.path, color: p.color}); 
                        particles.splice(i, 1); 
                        continue; 
                    }
                    
                    ctx.beginPath();
                    ctx.moveTo(p.path[0].x, p.path[0].y);
                    for (let j = 1; j < p.path.length; j++) {
                        ctx.lineTo(p.path[j].x, p.path[j].y);
                    }
                    ctx.strokeStyle = p.color;
                    ctx.lineWidth = 2;
                    ctx.setLineDash([5, 5]);
                    ctx.stroke();
                    ctx.setLineDash([]);
                    
                    ctx.beginPath();
                    ctx.arc(p.x, p.y, 6, 0, Math.PI * 2);
                    ctx.fillStyle = p.color;
                    ctx.fill();
                    ctx.strokeStyle = '#fff';
                    ctx.lineWidth = 1;
                    ctx.stroke();
                    
                    ctx.fillStyle = '#fff';
                    ctx.font = '10px Arial';
                    ctx.textAlign = 'center';
                    ctx.textBaseline = 'middle';
                    let sign = p.q > 0 ? '+' : (p.q < 0 ? '-' : '');
                    ctx.fillText(sign, p.x, p.y);
                }
                animationId = requestAnimationFrame(animate);
            }
            
            updateLabels();
            drawMagneticField();
        }
        
        initSimulation();
    })();
</script>
""".replace('UID', uid)

display(HTML(simulacion_html))
```
