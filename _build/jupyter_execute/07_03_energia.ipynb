
# 7.3 Energía Potencial Magnética

Como vimos en la sección anterior, una espira inmersa en un campo magnético uniforme experimenta un momento de torsión (torque) que tiende a alinear su momento dipolar magnético $\vec{\mu}$ con el campo $\vec{B}$. La presencia de este torque implica que existe un trabajo mecánico asociado a la rotación de la espira.

Recordemos de la mecánica clásica que el trabajo infinitesimal $dW$ realizado por un torque $\tau$ para rotar un objeto un ángulo $d\theta$ está dado por:

$$dW = -\tau d\theta$$

El signo negativo indica que el torque magnético realiza un trabajo que tiende a disminuir el ángulo $\theta$ entre $\vec{\mu}$ y $\vec{B}$ (es decir, busca la alineación).

Si sustituimos la magnitud del torque magnético, $\tau = \mu B \sin\theta$, en nuestra expresión de trabajo obtenemos:

$$dW = -(\mu B \sin\theta) d\theta$$

Al igual que en electrostática o en gravitación, el trabajo realizado por una fuerza o torque conservativo es igual al cambio negativo en la energía potencial ($dU = -dW$). Por lo tanto, el cambio infinitesimal en la energía potencial magnética $dU$ es:

$$dU = \mu B \sin\theta d\theta$$

Para encontrar la energía potencial macroscópica $U(\theta)$, integramos esta expresión. Por convención, definimos el nivel de referencia de la energía potencial ($U = 0$) cuando el momento magnético es perpendicular al campo magnético ($\theta = \pi/2$). Integramos entonces desde este ángulo de referencia hasta un ángulo arbitrario $\theta$:

$$U(\theta) - U(\pi/2) = \int_{\pi/2}^{\theta} \mu B \sin\theta' d\theta'$$

$$U(\theta) - 0 = \mu B \left[ -\cos\theta' \right]_{\pi/2}^{\theta}$$

$$U(\theta) = -\mu B \cos\theta$$

Reconociendo que $\mu B \cos\theta$ es la definición geométrica del producto escalar (o producto punto) entre dos vectores, obtenemos la expresión general para la energía potencial magnética de un dipolo:

$$U = -\vec{\mu} \cdot \vec{B}$$

## Análisis de los estados de energía

La ecuación $U = -\vec{\mu} \cdot \vec{B}$ nos permite identificar dos posiciones críticas para el momento dipolar en el campo magnético:

*   **Energía Mínima (Equilibrio Estable):** Ocurre cuando $\vec{\mu}$ y $\vec{B}$ son paralelos ($\theta = 0^\circ$). En este punto, $\cos(0) = 1$, por lo que $U_{\text{mín}} = -\mu B$. La espira está en su configuración de menor energía y cualquier pequeña perturbación generará un torque restaurador que la hará oscilar alrededor de este punto.
*   **Energía Máxima (Equilibrio Inestable):** Ocurre cuando $\vec{\mu}$ y $\vec{B}$ son antiparalelos ($\theta = 180^\circ$ o $\pi$). Aquí, $\cos(\pi) = -1$, por lo que $U_{\text{máx}} = +\mu B$. Aunque el torque neto en este punto exacto es cero, la espira está en el pico de su energía potencial; una perturbación infinitesimal provocará un torque que tenderá a girarla fuertemente hacia la posición estable.

Finalmente, el trabajo que debe realizar un agente externo para rotar el dipolo desde un ángulo inicial $\theta_i$ hasta un ángulo final $\theta_f$ a velocidad constante es igual a la diferencia de energía potencial:

$$W_{\text{ext}} = \Delta U = U_f - U_i = (-\vec{\mu} \cdot \vec{B}_f) - (-\vec{\mu} \cdot \vec{B}_i)$$

```python
import random
from IPython.display import display, HTML

uid_sim2 = str(random.randint(10000, 99999))

simulacion_energia_html = r"""
<div style="border: 1px solid #e0e0e0; padding: 20px; border-radius: 8px; background-color: #f8f9fa; font-family: -apple-system, BlinkMacSystemFont, 'Segoe UI', Roboto, Arial, sans-serif;">
    <h3 style="margin-top:0; color: #2c3e50;">Pozo de Energía Potencial</h3>
    <p style="font-size: 0.95em; color: #555; margin-bottom: 15px;">Fija un ángulo inicial y suelta la espira. Observa cómo oscila en el fondo del pozo (0°). Si partes de ±180°, el equilibrio inestable hará que caiga hacia el centro ante la mínima perturbación.</p>
    
    <div style="display: flex; gap: 20px; flex-wrap: wrap; margin-bottom: 15px;">
        <div style="flex: 1.2; min-width: 250px;">
            <div style="margin-bottom: 8px;">
                <label style="font-weight: 600; display: inline-block; width: 140px;">Ángulo Inicial (θ₀): <span id="valAng0_UID2" style="font-weight: normal;">120</span>°</label>
                <input type="range" id="Ang0_UID2" min="-180" max="180" step="1" value="120" style="width: 45%; vertical-align: middle;">
            </div>
            <div style="margin-bottom: 8px;">
                <label style="font-weight: 600; display: inline-block; width: 140px;">Campo (B): <span id="valB_UID2" style="font-weight: normal;">1.0</span> T</label>
                <input type="range" id="B_UID2" min="0.1" max="3" step="0.1" value="1.0" style="width: 45%; vertical-align: middle;">
            </div>
            <div style="margin-top: 15px; display: flex; gap: 10px;">
                <button id="btnPlay_UID2" style="flex: 1; padding: 8px; background-color: #0d6efd; color: white; border: none; border-radius: 4px; cursor: pointer; font-weight: bold;">▶ Soltar Espira</button>
                <button id="btnStop_UID2" style="flex: 1; padding: 8px; background-color: #6c757d; color: white; border: none; border-radius: 4px; cursor: pointer; font-weight: bold;">⏹ Detener</button>
            </div>
        </div>

        <div style="flex: 1; min-width: 220px; font-size: 0.95em; background: #fff; padding: 15px; border-radius: 6px; border: 1px solid #ccc; display: flex; flex-direction: column; justify-content: center;">
            <div style="margin-bottom: 5px; font-size: 1.1em;"><b>Ángulo Actual:</b> <span id="valTheta_UID2" style="color: #333; font-weight: bold;">120.0</span>°</div>
            <div style="margin-bottom: 5px; font-size: 1.1em;"><b>Torque (τ):</b> <span id="valTauDynamic_UID2" style="color: #dc3545; font-weight: bold;">0.00</span> N·m</div>
            <div style="margin-bottom: 5px; font-size: 1.1em;"><b>Energía (U):</b> <span id="valU_UID2" style="color: #6f42c1; font-weight: bold;">0.00</span> J</div>
        </div>
    </div>
    
    <div style="display: flex; gap: 10px; flex-wrap: wrap;">
        <!-- Vista superior de la espira -->
        <canvas id="cvs_sys_UID2" width="300" height="300" style="background: #ffffff; border: 1px solid #ccc; border-radius: 4px; flex: 1; min-width: 250px;"></canvas>
        <!-- Gráfica de Energía -->
        <canvas id="cvs_graph_UID2" width="450" height="300" style="background: #ffffff; border: 1px solid #ccc; border-radius: 4px; flex: 1.5; min-width: 300px;"></canvas>
    </div>
</div>

<script>
    (function() {
        setTimeout(function() {
            const cvsSys = document.getElementById('cvs_sys_UID2');
            const cvsGraph = document.getElementById('cvs_graph_UID2');
            if (!cvsSys || !cvsGraph) return;
            const ctxS = cvsSys.getContext('2d');
            const ctxG = cvsGraph.getContext('2d');
            
            const slider_Ang0 = document.getElementById('Ang0_UID2');
            const slider_B = document.getElementById('B_UID2');
            const btnPlay = document.getElementById('btnPlay_UID2');
            const btnStop = document.getElementById('btnStop_UID2');
            
            const val_Ang0 = document.getElementById('valAng0_UID2');
            const val_B = document.getElementById('valB_UID2');
            const val_Theta = document.getElementById('valTheta_UID2');
            const val_Tau = document.getElementById('valTauDynamic_UID2');
            const val_U = document.getElementById('valU_UID2');
            
            let isPlaying = false;
            let animationId;
            
            // Iniciar en 120 grados
            let theta = 120 * Math.PI / 180;
            let omega = 0;
            const mu = 2.0; 
            const I_inertia = 0.5; 
            const damping = 0.15; 
            let B_mag = 1.0;
            
            function drawArrow(ctx, x, y, dx, dy, color, lineWidth=2) {
                const headlen = 10; const angle = Math.atan2(dy, dx);
                ctx.beginPath(); ctx.moveTo(x, y); ctx.lineTo(x + dx, y + dy);
                ctx.strokeStyle = color; ctx.lineWidth = lineWidth; ctx.stroke();
                ctx.beginPath();
                ctx.moveTo(x + dx, y + dy);
                ctx.lineTo(x + dx - headlen * Math.cos(angle - Math.PI / 6), y + dy - headlen * Math.sin(angle - Math.PI / 6));
                ctx.lineTo(x + dx - headlen * Math.cos(angle + Math.PI / 6), y + dy - headlen * Math.sin(angle + Math.PI / 6));
                ctx.lineTo(x + dx, y + dy);
                ctx.fillStyle = color; ctx.fill();
            }

            function drawSystem() {
                ctxS.clearRect(0, 0, cvsSys.width, cvsSys.height);
                const cx = cvsSys.width / 2;
                const cy = cvsSys.height / 2;
                
                ctxS.strokeStyle = '#a0aabc'; ctxS.lineWidth = 1;
                for(let y = 30; y <= cvsSys.height; y += 40) {
                    drawArrow(ctxS, 10, y, cvsSys.width - 30, 0, '#a0aabc', 1.5);
                }
                ctxS.fillStyle = '#a0aabc'; ctxS.font = '14px Arial'; ctxS.fillText('Campo B', cvsSys.width - 65, 20);
                
                const loopRadius = 70;
                const loopX = loopRadius * Math.cos(theta + Math.PI/2);
                const loopY = loopRadius * Math.sin(theta + Math.PI/2);
                
                ctxS.beginPath();
                ctxS.moveTo(cx - loopX, cy + loopY);
                ctxS.lineTo(cx + loopX, cy - loopY);
                ctxS.strokeStyle = '#cd7f32'; ctxS.lineWidth = 6; ctxS.lineCap = 'round';
                ctxS.stroke();
                
                ctxS.beginPath(); ctxS.arc(cx, cy, 5, 0, Math.PI*2);
                ctxS.fillStyle = '#333'; ctxS.fill();
                
                const muLen = 80;
                const mux = muLen * Math.cos(theta);
                const muy = muLen * -Math.sin(theta);
                drawArrow(ctxS, cx, cy, mux, muy, '#0d6efd', 4);
                ctxS.fillStyle = '#0d6efd'; ctxS.font = 'bold 16px sans-serif';
                ctxS.fillText('μ', cx + mux + 10, cy + muy + 10);
            }

            function drawEnergyGraph() {
                ctxG.clearRect(0, 0, cvsGraph.width, cvsGraph.height);
                const w = cvsGraph.width; const h = cvsGraph.height;
                const pX = 45; const pY = 25; const pW = w - 65; const pH = h - 65;
                
                ctxG.strokeStyle = '#333'; ctxG.lineWidth = 1.5;
                const zeroY = pY + pH/2;
                
                ctxG.beginPath(); ctxG.moveTo(pX, pY); ctxG.lineTo(pX, pY + pH); ctxG.stroke();
                ctxG.beginPath(); ctxG.moveTo(pX, zeroY); ctxG.lineTo(pX + pW, zeroY); ctxG.stroke();
                
                ctxG.fillStyle = '#555'; ctxG.font = 'bold 13px sans-serif';
                ctxG.fillText('U(θ)', 5, pY - 8); 
                ctxG.font = '12px sans-serif';
                ctxG.fillText('0', 25, zeroY + 4);
                ctxG.fillText('+μB', 15, pY + 12); 
                ctxG.fillText('-μB', 18, pY + pH);
                
                ctxG.fillText('-180°', pX - 10, zeroY + 20);
                ctxG.fillText('0°', pX + pW/2 - 5, zeroY + 20);
                ctxG.fillText('+180°', pX + pW - 15, zeroY + 20);
                
                ctxG.beginPath();
                ctxG.strokeStyle = '#6f42c1'; ctxG.lineWidth = 2.5;
                for(let a = -180; a <= 180; a++) {
                    let rad = a * Math.PI / 180;
                    let u_val = -mu * B_mag * Math.cos(rad);
                    let gx = pX + ((a + 180) / 360) * pW;
                    let gy = zeroY - (u_val / (mu * B_mag)) * (pH/2); 
                    if(a === -180) ctxG.moveTo(gx, gy);
                    else ctxG.lineTo(gx, gy);
                }
                ctxG.stroke();
                
                let normTheta = theta % (2*Math.PI);
                if (normTheta > Math.PI) normTheta -= 2*Math.PI;
                if (normTheta < -Math.PI) normTheta += 2*Math.PI;
                
                let current_U = -mu * B_mag * Math.cos(normTheta);
                let pointX = pX + ((normTheta + Math.PI) / (2*Math.PI)) * pW;
                let pointY = zeroY - (current_U / (mu * B_mag)) * (pH/2);
                
                ctxG.beginPath(); ctxG.arc(pointX, pointY, 6, 0, Math.PI*2);
                ctxG.fillStyle = '#dc3545'; ctxG.fill();
                ctxG.strokeStyle = '#fff'; ctxG.lineWidth = 2; ctxG.stroke();
                
                ctxG.beginPath(); ctxG.moveTo(pointX, zeroY); ctxG.lineTo(pointX, pointY);
                ctxG.strokeStyle = '#dc3545'; ctxG.lineWidth = 1; ctxG.setLineDash([4, 4]); ctxG.stroke(); ctxG.setLineDash([]);
            }

            function updateUI() {
                val_B.innerText = B_mag.toFixed(1);
                
                let normTheta = theta % (2*Math.PI);
                if (normTheta > Math.PI) normTheta -= 2*Math.PI;
                if (normTheta < -Math.PI) normTheta += 2*Math.PI;
                
                let degTheta = normTheta * 180 / Math.PI;
                val_Theta.innerText = degTheta.toFixed(1);
                
                let tau = -mu * B_mag * Math.sin(theta);
                val_Tau.innerText = Math.abs(tau).toFixed(2);
                
                let energy = -mu * B_mag * Math.cos(theta);
                val_U.innerText = energy.toFixed(2);
            }

            function physicsLoop() {
                if(isPlaying) {
                    const dt = 0.04;
                    
                    let normTheta = theta % (2*Math.PI);
                    if (normTheta > Math.PI) normTheta -= 2*Math.PI;
                    if (normTheta < -Math.PI) normTheta += 2*Math.PI;
                    
                    let normDeg = normTheta * 180 / Math.PI;
                    
                    if (Math.abs(Math.abs(normDeg) - 180) < 0.1 && omega === 0) {
                        if (normDeg >= 179.9) theta -= 0.01;
                        else if (normDeg <= -179.9) theta += 0.01;
                    }
                    
                    let torque = -mu * B_mag * Math.sin(theta);
                    let alpha = (torque - damping * omega) / I_inertia;
                    
                    omega += alpha * dt;
                    theta += omega * dt;
                }
                
                drawSystem();
                drawEnergyGraph();
                updateUI();
                
                if(isPlaying) {
                    animationId = requestAnimationFrame(physicsLoop);
                }
            }

            slider_Ang0.addEventListener('input', () => {
                isPlaying = false;
                cancelAnimationFrame(animationId);
                btnPlay.innerText = "▶ Soltar Espira";
                val_Ang0.innerText = slider_Ang0.value;
                theta = parseFloat(slider_Ang0.value) * Math.PI / 180;
                omega = 0; 
                physicsLoop(); 
            });
            
            slider_B.addEventListener('input', () => {
                B_mag = parseFloat(slider_B.value);
                if(!isPlaying) physicsLoop();
            });
            
            btnPlay.addEventListener('click', () => {
                if(!isPlaying) {
                    isPlaying = true;
                    btnPlay.innerText = "En movimiento...";
                    physicsLoop();
                }
            });
            
            btnStop.addEventListener('click', () => {
                isPlaying = false;
                cancelAnimationFrame(animationId);
                btnPlay.innerText = "▶ Soltar Espira";
                theta = parseFloat(slider_Ang0.value) * Math.PI / 180;
                omega = 0;
                physicsLoop();
            });
            
            B_mag = parseFloat(slider_B.value);
            physicsLoop();
            
        }, 150);
    })();
</script>
""".replace('UID2', uid_sim2)

display(HTML(simulacion_energia_html))
```
