# Circuitos RC: Comportamiento Transitorio

Los circuitos RC (Resistencia-Condensador) se caracterizan porque la corriente y la carga dependen del tiempo. Este comportamiento es fundamental para entender desde temporizadores hasta el filtrado de señales.

## 1. Carga de un Condensador

Consideremos un circuito con una fuente $V_0$, una resistencia $R$ y un condensador $C$ inicialmente descargado. Al cerrar el interruptor en $t=0$, el condensador comienza a acumular carga.

La ecuación de malla para este circuito es:

$$V_0 - IR - \frac{q}{C} = 0$$

Como $I = \frac{dq}{dt}$, obtenemos la siguiente ecuación diferencial:

$$V_0 = R\frac{dq}{dt} + \frac{q}{C}$$

### Solución de la Ecuación

Resolviendo mediante integración, la carga en función del tiempo es:

$$q(t) = V_0 C (1 - e^{-t/RC})$$

Donde definimos la **constante de tiempo** $\tau = RC$. Esta constante representa el tiempo necesario para que el condensador alcance aproximadamente el 63.2% de su carga máxima ($Q_0 = V_0 C$).

La corriente en el circuito disminuye exponencialmente:

$$I(t) = \frac{dq}{dt} = \frac{V_0}{R} e^{-t/RC}$$

## 2. Descarga de un Condensador

Si un condensador con carga inicial $Q$ se desconecta de la fuente y se cierra sobre una resistencia $R$, la energía almacenada se disipa en forma de calor.

La ecuación del circuito es:

$$IR + \frac{q}{C} = 0 \implies R\frac{dq}{dt} + \frac{q}{C} = 0$$

La solución para la carga y la corriente durante la descarga es:
* **Carga:** $q(t) = Q e^{-t/RC}$
* **Corriente:** $I(t) = -\frac{V_0}{R} e^{-t/RC}$ (El signo negativo indica que la corriente fluye en sentido opuesto a la carga).

## 3. Análisis en Tiempos Límite

Para circuitos complejos, es útil analizar el comportamiento en los extremos temporales:
* **En $t = 0$ (Instante inicial):** Un condensador descargado se comporta como un **alambre** (cortocircuito), permitiendo el flujo máximo de corriente.
* **En $t \to \infty$ (Estado estacionario):** El condensador se comporta como un **circuito abierto**, interrumpiendo la corriente en su rama una vez que está totalmente cargado.

### Ejemplo de Aplicación

En un circuito donde el condensador está en paralelo con otras resistencias, la constante de tiempo equivalente puede variar. Por ejemplo, para el circuito analizado en clase con una resistencia $R$ en serie y una combinación paralelo, se determinó que:

$$\tau_{eq} = \frac{2}{3}RC$$

$$q_{max} = \frac{2}{3}CV_0$$
