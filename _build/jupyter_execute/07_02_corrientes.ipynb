# 7.2 Fuerzas sobre Corrientes

¿Qué pasa cuando hay muchas cargas en movimiento? Analizaremos la fuerza magnética sobre un conductor que lleva corriente.

La fuerza magnética para una sola carga en movimiento está dada por:

$$\vec{F} = q \, \vec{v}_d \times \vec{B}$$

Considerando todas las cargas en el volumen, la fuerza total es:

$$\vec{F} = n q \vec{v}_d A l \times \vec{B} = (n q v_d A) \vec{l} \times \vec{B}$$

Donde:
* $N$ es el número de partículas.
* $V$ es el volumen del conductor, tal que $V = Al$ ($A$ es el área transversal y $l$ la longitud).
* $n = \frac{N}{V}$ es la densidad de partículas.

Recordando que la corriente eléctrica se define como $I = n q v_d A$, obtenemos la expresión para la fuerza magnética sobre un alambre recto de longitud $l$:

$$\vec{F} = I \, \vec{l} \times \vec{B}$$

Por ejemplo, si tenemos un alambre sobre el eje $y$ con una corriente $I$ y un campo magnético uniforme apuntando hacia adentro de la página ($\vec{B} = - B\, \hat{k}$):

$$\vec{F} = I l \hat{j} \times B(-\hat{k}) = -I l B \, \hat{\imath}$$

## Expresión general para un alambre de forma arbitraria

Para un conductor de forma arbitraria, podemos tomar un elemento diferencial de longitud $d\vec{l}$. La fuerza sobre este pequeño segmento es:

$$d\vec{F} = I d\vec{l} \times \vec{B}$$

Integrando a lo largo del conductor desde un punto inicial $i$ hasta un punto final $f$:

$$\vec{F} = \int_{i}^{f} I d\vec{l} \times \vec{B}$$

**Si el campo magnético $\vec{B}$ es constante**, puede salir de la integral:

$$\vec{F} = I \left( \int_{i}^{f} d\vec{l} \right) \times \vec{B}$$

Una **espira** es un alambre cerrado. En este caso, el punto inicial coincide con el punto final ($i=f$), por lo que $\int d\vec{l} = \vec{0}$. En consecuencia:

La fuerza neta sobre una espira conductora cerrada, inmersa en un campo magnético constante, es igual a cero ($\vec{F} = \vec{0}$).

### Ejemplo: Alambre semicircular
Para un alambre semicircular de radio $R$ en el plano $xy$ bajo un campo $\vec{B} = B(-\hat{k})$:

$$\vec{F} = I \left( \int d\vec{l} \right) \times \vec{B} = I \int_{0}^{\pi} (-\sin\theta \, \hat{\imath} + \cos\theta \, \hat{\jmath}) R d\theta \times B(-\hat{k})$$

### Ejemplo: Espira cuadrada
Consideremos una espira cuadrada de lado $a$. La fuerza neta es la suma de las fuerzas sobre cada segmento: $\vec{F} = \vec{F}_1 + \vec{F}_2 + \vec{F}_3 + \vec{F}_4$. Para un campo magnético particular, las fuerzas opuestas se cancelan. Por ejemplo, $\vec{F}_1 = IaB(-\hat{k})$ y $\vec{F}_3 = IaB(+\hat{k})$, resultando nuevamente en $\vec{F} = \vec{0}$.

## Torque sobre una espira

Aunque la fuerza neta sobre una espira en un campo uniforme es cero, las fuerzas magnéticas pueden ejercer un momento de torsión (torque) que hace girar la espira.

Considerando la espira cuadrada de lado $a$, el torque neto respecto al centro se calcula sumando los torques individuales $\vec{\tau}_i = \vec{r}_i \times \vec{F}_i$:

$$\vec{\tau} = \vec{\tau}_1 + \vec{\tau}_2 + \vec{\tau}_3 + \vec{\tau}_4 = \vec{r}_1 \times \vec{F}_1 + \vec{r}_2 \times \vec{F}_2$$

Evaluando para las fuerzas no nulas en un campo magnético:

$$\vec{\tau} = \frac{a}{2} \hat{i} \times I a B (-\hat{k}) + \frac{a}{2} (-\hat{\imath}) \times I a B \, \hat{k} = \frac{1}{2} a^2 I B \, \hat{\jmath} + \frac{1}{2} a^2 I B \, \hat{\jmath} = I a^2 B \, \hat{\jmath}$$

Reconociendo que el área de la espira es $A = a^2$, podemos generalizar el resultado mediante un producto cruz:

$$\vec{\tau} = I \vec{A} \times \vec{B}$$

Definimos el **momento magnético** de la espira como:

$$\vec{\mu} = I \vec{A}$$

Por lo tanto, el torque magnético sobre cualquier espira plana se escribe en su forma general como:

$$\vec{\tau} = \vec{\mu} \times \vec{B}$$
