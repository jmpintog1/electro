# 7.2 Fuerzas sobre Corrientes

¿Qué pasa cuando hay muchas cargas en movimiento? Analizaremos la fuerza magnética sobre un conductor que lleva corriente.

La fuerza magnética para una sola carga en movimiento está dada por:

$$\vec{F} = q \, \vec{v}_d \times \vec{B}$$

Considerando todas las cargas en el volumen, la fuerza total es:

$$\vec{F} = n q \vec{v}_d A l \times \vec{B} = (n q v_d A) \vec{l} \times \vec{B}$$

Donde:
* $N$ es el número de partículas.
* $V$ es el volumen del conductor, tal que $V = Al$ ($A$ es el área transversal y $l$ la longitud).
* $n = \frac{N}{V}$ es la densidad de partículas.

Recordando que la corriente eléctrica se define como $I = n q v_d A$, obtenemos la expresión para la fuerza magnética sobre un alambre recto de longitud $l$:

$$\vec{F} = I \, \vec{l} \times \vec{B}$$

Por ejemplo, si tenemos un alambre sobre el eje $y$ con una corriente $I$ y un campo magnético uniforme apuntando hacia adentro de la página ($\vec{B} = - B\, \hat{k}$):

$$\vec{F} = I l \, \hat{\jmath} \times B(-\hat{k}) = -I l B \, \hat{\imath}$$

## Expresión general para un alambre de forma arbitraria

Para un conductor de forma arbitraria, podemos tomar un elemento diferencial de longitud $d\vec{l}$. La fuerza sobre este pequeño segmento es:

$$d\vec{F} = I d\vec{l} \times \vec{B}$$

Integrando a lo largo del conductor desde un punto inicial $i$ hasta un punto final $f$:

$$\vec{F} = \int_{i}^{f} I d\vec{l} \times \vec{B}$$

**Si el campo magnético $\vec{B}$ es constante**, puede salir de la integral:

$$\vec{F} = I \left( \int_{i}^{f} d\vec{l} \right) \times \vec{B}$$

Una **espira** es un alambre cerrado. En este caso, el punto inicial coincide con el punto final ($i=f$), por lo que $\int d\vec{l} = \vec{0}$. En consecuencia:

La fuerza neta sobre una espira conductora cerrada, inmersa en un campo magnético constante, es igual a cero ($\vec{F} = \vec{0}$).

### Ejemplo: Alambre semicircular
Para un alambre semicircular de radio $R$ en el plano $xy$ bajo un campo $\vec{B} = B(-\hat{k})$:

$$\vec{F} = I \left( \int d\vec{l} \right) \times \vec{B} = I \int_{0}^{\pi} (-\sin\theta \, \hat{\imath} + \cos\theta \, \hat{\jmath}) R d\theta \times B(-\hat{k})$$

### Ejemplo: Espira cuadrada
Consideremos una espira cuadrada de lado $a$. La fuerza neta es la suma de las fuerzas sobre cada segmento: $\vec{F} = \vec{F}_1 + \vec{F}_2 + \vec{F}_3 + \vec{F}_4$. Para un campo magnético particular, las fuerzas opuestas se cancelan. Por ejemplo, $\vec{F}_1 = IaB(-\hat{k})$ y $\vec{F}_3 = IaB(+\hat{k})$, resultando nuevamente en $\vec{F} = \vec{0}$.

## Torque sobre una espira

Aunque la fuerza neta sobre una espira en un campo uniforme es cero, las fuerzas magnéticas pueden ejercer un momento de torsión (torque) que hace girar la espira.

Considerando la espira cuadrada de lado $a$, el torque neto respecto al centro se calcula sumando los torques individuales $\vec{\tau}_i = \vec{r}_i \times \vec{F}_i$:

$$\vec{\tau} = \vec{\tau}_1 + \vec{\tau}_2 + \vec{\tau}_3 + \vec{\tau}_4 = \vec{r}_1 \times \vec{F}_1 + \vec{r}_2 \times \vec{F}_2$$

Evaluando para las fuerzas no nulas en un campo magnético:

$$\vec{\tau} = \frac{a}{2} \hat{i} \times I a B (-\hat{k}) + \frac{a}{2} (-\hat{\imath}) \times I a B \, \hat{k} = \frac{1}{2} a^2 I B \, \hat{\jmath} + \frac{1}{2} a^2 I B \, \hat{\jmath} = I a^2 B \, \hat{\jmath}$$

Reconociendo que el área de la espira es $A = a^2$, podemos generalizar el resultado mediante un producto cruz:

$$\vec{\tau} = I \vec{A} \times \vec{B}$$

Definimos el **momento magnético** de la espira como:

$$\vec{\mu} = I \vec{A}$$

Por lo tanto, el torque magnético sobre cualquier espira plana se escribe en su forma general como:

$$\vec{\tau} = \vec{\mu} \times \vec{B}$$

```python
import random
from IPython.display import display, HTML

uid_sim5 = str(random.randint(10000, 99999))

simulacion_dual_html = r"""
<div style="border: 1px solid #e0e0e0; padding: 20px; border-radius: 8px; background-color: #f8f9fa; font-family: -apple-system, BlinkMacSystemFont, 'Segoe UI', Roboto, Arial, sans-serif;">
    <h3 style="margin-top:0; color: #2c3e50;">Perspectivas 2D: Fuerzas y Torque</h3>
    <p style="font-size: 0.95em; color: #555; margin-bottom: 15px;">Alterna entre las vistas. En la <b>Vista Frontal</b>, el campo entra a la pantalla y las fuerzas solo expanden la espira ($\tau=0$). En la <b>Vista Superior</b>, se revela el desfase de las fuerzas que genera el giro.</p>
    
    <div style="display: flex; gap: 20px; flex-wrap: wrap; margin-bottom: 15px;">
        <div style="flex: 1.2; min-width: 250px;">
            <div style="margin-bottom: 8px;">
                <label style="font-weight: 600; display: inline-block; width: 120px;">Corriente (I):</label>
                <input type="range" id="I_UID5" min="1" max="10" step="0.5" value="5.0" style="width: 45%; vertical-align: middle;">
            </div>
            <div style="margin-bottom: 8px;">
                <label style="font-weight: 600; display: inline-block; width: 120px;">Campo (B):</label>
                <input type="range" id="B_UID5" min="0.5" max="5" step="0.5" value="2.0" style="width: 45%; vertical-align: middle;">
            </div>
            <div style="margin-bottom: 8px;">
                <label style="font-weight: 600; display: inline-block; width: 120px;">Ángulo (θ): <span id="valAng_UID5" style="font-weight: normal;">45</span>°</label>
                <input type="range" id="Ang_UID5" min="0" max="360" step="1" value="45" style="width: 45%; vertical-align: middle;">
            </div>
            <div style="margin-top: 15px; display: flex; gap: 10px;">
                <button id="btnFrontal_UID5" style="flex: 1; padding: 8px; background-color: #6c757d; color: white; border: none; border-radius: 4px; cursor: pointer; font-weight: bold;">[1] Vista Frontal</button>
                <button id="btnSuperior_UID5" style="flex: 1; padding: 8px; background-color: #0d6efd; color: white; border: none; border-radius: 4px; cursor: pointer; font-weight: bold;">[2] Vista Superior</button>
            </div>
        </div>

        <div style="flex: 1; min-width: 220px; font-size: 1.0em; background: #fff; padding: 15px; border-radius: 6px; border: 1px solid #ccc; display: flex; flex-direction: column; justify-content: center;">
            <div style="margin-bottom: 10px; font-size: 1.1em;"><b>Fuerza Neta ($\sum \vec{F}$):</b> <span style="color: #198754; font-weight: bold;">0.00 N</span></div>
            <div style="margin-bottom: 10px; font-size: 1.1em;"><b>Torque ($\vec{\tau}$):</b> <span id="valTau_UID5" style="color: #6f42c1; font-weight: bold;">0.00</span> N·m</div>
            <div id="txtExplicacion_UID5" style="font-size: 0.85em; color: #dc3545; font-style: italic;">En esta vista, el torque es 0 porque todas las fuerzas empujan desde el centro.</div>
        </div>
    </div>
    
    <div style="text-align: center; position: relative;">
        <canvas id="canvas_UID5" width="700" height="380" style="background: #ffffff; border: 1px solid #ccc; border-radius: 4px; box-shadow: 0 2px 5px rgba(0,0,0,0.05); max-width: 100%;"></canvas>
    </div>
</div>

<script>
    (function() {
        setTimeout(function() {
            const canvas = document.getElementById('canvas_UID5');
            if (!canvas) return;
            const ctx = canvas.getContext('2d');
            
            const in_I = document.getElementById('I_UID5');
            const in_B = document.getElementById('B_UID5');
            const in_Ang = document.getElementById('Ang_UID5');
            const val_Ang = document.getElementById('valAng_UID5');
            const val_Tau = document.getElementById('valTau_UID5');
            const txtExpl = document.getElementById('txtExplicacion_UID5');
            
            const btnFrontal = document.getElementById('btnFrontal_UID5');
            const btnSuperior = document.getElementById('btnSuperior_UID5');
            
            let viewMode = 'superior'; // 'frontal' o 'superior'
            const area = 0.04; // m^2 (Ej: 20cm x 20cm)
            const w_px = 160; 
            const h_px = 160;

            function switchView(mode) {
                viewMode = mode;
                if(mode === 'frontal') {
                    btnFrontal.style.backgroundColor = '#0d6efd';
                    btnSuperior.style.backgroundColor = '#6c757d';
                    txtExpl.innerText = "Las fuerzas tiran hacia afuera (expansión). La línea de acción pasa por el centro: Torque = 0.";
                } else {
                    btnSuperior.style.backgroundColor = '#0d6efd';
                    btnFrontal.style.backgroundColor = '#6c757d';
                    txtExpl.innerText = "Las fuerzas laterales están desfasadas respecto al centro. ¡Se genera el par de fuerzas que causa el Torque!";
                }
                render();
            }

            function drawArrow(x, y, dx, dy, color, width=2, label='') {
                const headlen = 10;
                const angle = Math.atan2(dy, dx);
                ctx.beginPath(); ctx.moveTo(x, y); ctx.lineTo(x + dx, y + dy);
                ctx.strokeStyle = color; ctx.lineWidth = width; ctx.stroke();
                
                ctx.beginPath();
                ctx.moveTo(x + dx, y + dy);
                ctx.lineTo(x + dx - headlen * Math.cos(angle - Math.PI / 6), y + dy - headlen * Math.sin(angle - Math.PI / 6));
                ctx.lineTo(x + dx - headlen * Math.cos(angle + Math.PI / 6), y + dy - headlen * Math.sin(angle + Math.PI / 6));
                ctx.lineTo(x + dx, y + dy);
                ctx.fillStyle = color; ctx.fill();
                
                if(label) {
                    ctx.font = 'bold 14px sans-serif';
                    ctx.fillText(label, x + dx + 8*Math.cos(angle), y + dy + 8*Math.sin(angle));
                }
            }

            function drawCurrentArrow(x1, y1, x2, y2) {
                const mx = (x1 + x2) / 2; const my = (y1 + y2) / 2;
                const angle = Math.atan2(y2 - y1, x2 - x1);
                const hl = 8;
                ctx.beginPath();
                ctx.moveTo(mx + (hl/2) * Math.cos(angle), my + (hl/2) * Math.sin(angle));
                ctx.lineTo(mx - hl * Math.cos(angle - Math.PI / 6), my - hl * Math.sin(angle - Math.PI / 6));
                ctx.lineTo(mx - hl * Math.cos(angle + Math.PI / 6), my - hl * Math.sin(angle + Math.PI / 6));
                ctx.closePath(); ctx.fillStyle = '#111'; ctx.fill();
            }

            function render() {
                ctx.clearRect(0, 0, canvas.width, canvas.height);
                const cx = canvas.width / 2;
                const cy = canvas.height / 2;
                
                const I = parseFloat(in_I.value);
                const B = parseFloat(in_B.value);
                const theta_deg = parseFloat(in_Ang.value);
                const theta = theta_deg * Math.PI / 180;
                
                val_Ang.innerText = theta_deg;
                
                const f_scale = 10 * B * (I/5); // Factor visual para las flechas

                if (viewMode === 'frontal') {
                    // --- VISTA FRONTAL ---
                    val_Tau.innerText = "0.00"; // En esta vista no hay torque
                    
                    // Fondo: B entrando (Cruces)
                    ctx.fillStyle = '#a0aabc'; ctx.font = '14px sans-serif';
                    for(let x=20; x<canvas.width; x+=40) {
                        for(let y=20; y<canvas.height; y+=40) {
                            ctx.fillText('×', x, y);
                        }
                    }
                    ctx.font = 'bold 16px sans-serif'; ctx.fillText('Campo B (Entrando ⊗)', 20, 30);
                    
                    // Rotar y dibujar rectángulo
                    ctx.save();
                    ctx.translate(cx, cy);
                    ctx.rotate(-theta); // Rotación visual 2D
                    
                    ctx.strokeStyle = '#cd7f32'; ctx.lineWidth = 4;
                    ctx.strokeRect(-w_px/2, -h_px/2, w_px, h_px);
                    
                    drawCurrentArrow(-w_px/2, -h_px/2, w_px/2, -h_px/2); // Top
                    drawCurrentArrow(w_px/2, -h_px/2, w_px/2, h_px/2);   // Right
                    drawCurrentArrow(w_px/2, h_px/2, -w_px/2, h_px/2);   // Bottom
                    drawCurrentArrow(-w_px/2, h_px/2, -w_px/2, -h_px/2); // Left
                    
                    // Fuerzas (siempre perpendiculares y hacia afuera por F = I x B)
                    drawArrow(0, -h_px/2, 0, -f_scale*2, '#dc3545', 3, 'F'); // Arriba
                    drawArrow(0, h_px/2, 0, f_scale*2, '#dc3545', 3, 'F');   // Abajo
                    drawArrow(w_px/2, 0, f_scale*2, 0, '#dc3545', 3, 'F');   // Derecha
                    drawArrow(-w_px/2, 0, -f_scale*2, 0, '#dc3545', 3, 'F'); // Izquierda
                    
                    // Vector Momento (Saliendo de la pantalla)
                    ctx.beginPath(); ctx.arc(0, 0, 8, 0, Math.PI*2); 
                    ctx.fillStyle = '#0d6efd'; ctx.fill();
                    ctx.fillStyle = '#fff'; ctx.fillText('·', -2, 4); // Punto
                    ctx.fillStyle = '#0d6efd'; ctx.font = 'bold 14px sans-serif'; ctx.fillText('μ (Saliendo)', 12, -10);
                    
                    ctx.restore();

                } else {
                    // --- VISTA SUPERIOR ---
                    const tau = I * area * B * Math.sin(theta);
                    val_Tau.innerText = Math.abs(tau).toFixed(2);
                    
                    // Fondo: B hacia la derecha
                    ctx.strokeStyle = '#a0aabc'; ctx.lineWidth = 1.5; ctx.setLineDash([10,10]);
                    for(let y=40; y<canvas.height; y+=50) {
                        drawArrow(10, y, canvas.width - 40, 0, '#a0aabc', 1.5);
                    }
                    ctx.setLineDash([]);
                    ctx.fillStyle = '#a0aabc'; ctx.font = 'bold 16px sans-serif'; ctx.fillText('Campo B (Horizontal →)', 20, 30);
                    
                    // La espira vista desde arriba es una varilla. 
                    // Si theta = 0, mu apunta a +X, varilla es vertical.
                    const alpha = theta + Math.PI/2;
                    
                    const x1 = cx + (w_px/2) * Math.cos(alpha);
                    const y1 = cy - (w_px/2) * Math.sin(alpha); // Invertido en Canvas
                    
                    const x2 = cx - (w_px/2) * Math.cos(alpha);
                    const y2 = cy + (w_px/2) * Math.sin(alpha);
                    
                    // Pivote
                    ctx.beginPath(); ctx.arc(cx, cy, 4, 0, Math.PI*2); ctx.fillStyle = '#333'; ctx.fill();
                    
                    // Varilla
                    ctx.strokeStyle = '#cd7f32'; ctx.lineWidth = 6; ctx.lineCap = 'round';
                    ctx.beginPath(); ctx.moveTo(x1, y1); ctx.lineTo(x2, y2); ctx.stroke();
                    
                    // Indicadores de Corriente (Saliente y Entrante)
                    // P1 (Corriente saliendo)
                    ctx.beginPath(); ctx.arc(x1, y1, 8, 0, Math.PI*2); ctx.fillStyle = '#fff'; ctx.fill();
                    ctx.lineWidth = 2; ctx.strokeStyle = '#111'; ctx.stroke();
                    ctx.fillStyle = '#111'; ctx.font = '14px sans-serif'; ctx.fillText('·', x1-3, y1+4);
                    
                    // P2 (Corriente entrando)
                    ctx.beginPath(); ctx.arc(x2, y2, 8, 0, Math.PI*2); ctx.fillStyle = '#fff'; ctx.fill();
                    ctx.lineWidth = 2; ctx.strokeStyle = '#111'; ctx.stroke();
                    ctx.fillStyle = '#111'; ctx.fillText('×', x2-4, y2+4);

                    // Momento dipolar (mu)
                    const muLen = 80;
                    drawArrow(cx, cy, muLen*Math.cos(theta), -muLen*Math.sin(theta), '#0d6efd', 3, 'μ');

                    // Fuerzas (P1 sufre F hacia arriba, P2 sufre F hacia abajo)
                    drawArrow(x1, y1 - 10, 0, -f_scale*2.5, '#dc3545', 3, 'F'); // Arriba (en canvas -y)
                    drawArrow(x2, y2 + 10, 0, f_scale*2.5, '#dc3545', 3, 'F');  // Abajo (en canvas +y)
                }
            }

            in_I.addEventListener('input', render);
            in_B.addEventListener('input', render);
            in_Ang.addEventListener('input', render);
            
            btnFrontal.addEventListener('click', () => switchView('frontal'));
            btnSuperior.addEventListener('click', () => switchView('superior'));
            
            switchView('superior'); // Iniciar en vista superior
        }, 150);
    })();
</script>
""".replace('UID5', uid_sim5)

display(HTML(simulacion_dual_html))
```
