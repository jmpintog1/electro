
# 9.1 Flujo Magnético

Para comprender rigurosamente los fenómenos de inducción electromagnética, es fundamental establecer el concepto de **flujo magnético**. De manera análoga al flujo eléctrico, el flujo magnético proporciona una medida cuantitativa de la cantidad de líneas de campo magnético que atraviesan una superficie determinada.

Matemáticamente, el flujo magnético, denotado por $\Phi_{m}$, a través de una superficie abierta o cerrada se define como la integral de superficie del campo magnético $\vec{B}$ que la atraviesa:

$$\Phi_{m}=\int\vec{B}\cdot d\vec{s}$$

Donde $d\vec{s}=ds\hat{n}$ es el vector diferencial de área, cuya magnitud es $ds$ y su dirección es perpendicular a la superficie en cada punto (definida por el vector unitario normal $\hat{n}$).

## Casos Particulares Importantes

**1. Superficies Cerradas:**
Si la superficie evaluada es cerrada (como una esfera o un cilindro tridimensional), el flujo magnético neto siempre es cero. Esto es una consecuencia directa de la inexistencia de monopolos magnéticos aislados en la naturaleza; toda línea de campo que entra a la superficie debe salir de ella:

$$\oint\vec{B}\cdot d\vec{s}=0$$

**2. Campo Magnético Uniforme y Superficie Plana:**
Cuando el campo magnético $\vec{B}$ es constante (uniforme en magnitud y dirección en todos los puntos de la superficie) y la superficie es plana, la integral se simplifica de manera considerable. La expresión se reduce a evaluar el ángulo $\theta$ entre el campo magnético y el vector normal a la superficie:

$$\Phi_{m}=\int B\hat{n}\cdot ds\hat{n}=\int B\cos\theta ds$$

Dado que $B$ y $\theta$ son constantes, se extraen de la integral, resultando en:

$$\Phi_{m}=BA\cos\theta$$

## Ejemplos de Cálculo de Flujo

**Ejemplo 1: Campo uniforme a través de un área rectangular**
Consideremos una espira rectangular de lados $a$ y $b$, de modo que su área total es $A=ab$. Si un campo magnético constante atraviesa esta espira de forma completamente perpendicular a su plano, los vectores $\vec{B}$ y $\hat{n}$ son paralelos, por lo que el ángulo es $0^{\circ}$ y $\cos(0^{\circ})=1$. El flujo a través de esta espira se define como:

$$\Phi_{m}=BA=Bab$$

**Ejemplo 2: Campo variable generado por un alambre infinito**
Un caso más complejo ocurre cuando el campo magnético no es espacialmente constante. Supongamos un alambre recto, largo e infinito que transporta una corriente $I$. A una distancia $d$ del alambre se ubica una espira rectangular de ancho $a$ y largo $b$.

De acuerdo con la ley de Ampère, el campo magnético generado por el alambre a una distancia transversal $x$ varía inversamente con la distancia:

$$\vec{B}=-B\hat{k}=\frac{\mu_{0}I}{2\pi x}(-\hat{k})$$

El diferencial de área de la espira orientada en la misma dirección es $d\vec{s}=ds(-\hat{k})$. Al calcular el flujo, es necesario integrar sobre el área de la espira tomando diferenciales bidimensionales $ds=dxdy$, ya que el campo depende de la posición $x$:

$$\Phi_{m}=\int\frac{\mu_{0}I}{2\pi x}(-\hat{k})\cdot ds(-\hat{k})$$

Agrupando las constantes e integrando en las dimensiones de la espira (desde $x=d$ hasta $x=d+a$ a lo ancho, y desde $y=0$ hasta $y=b$ a lo largo):

$$\Phi_{m}=\frac{\mu_{0}I}{2\pi}\int_{d}^{d+a}\frac{dx}{x}\int_{0}^{b}dy$$

Evaluando estas integrales definidas, se obtiene el flujo magnético total sobre la espira:

$$\Phi_{m}=\frac{\mu_{0}I}{2\pi}\ln\left(\frac{d+a}{d}\right)b$$

## Unidades y Relación con la Inducción

En el Sistema Internacional, la unidad fundamental de medida del flujo magnético es el **Weber (Wb)**. Al observar la ecuación general, queda en evidencia que el flujo es el producto de un campo magnético por un área, lo que nos da la equivalencia:

**1 Wb = 1 T·m²**

La importancia física del flujo magnético se hace patente en situaciones dinámicas. Según la **Ley de Faraday**, la variación temporal del flujo magnético a través de un circuito induce en él una fuerza electromotriz ($fem$ o $\Delta V_{ind}$), equivalente a un voltaje:

$$\Delta V_{ind}=-\frac{d\Phi_{m}}{dt}$$

Esta relación diferencial también define una importante equivalencia de unidades donde **1 Wb = 1 V·s**. Esto nos indica directamente que una tasa de cambio de flujo de **1 Wb/s** genera en el circuito una diferencia de potencial inducida de **1 Volt**.
