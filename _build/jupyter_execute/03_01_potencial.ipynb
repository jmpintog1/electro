# 3.1 Campo y Potencial Eléctrico

Para comenzar nuestro estudio, recordemos que el campo electrostático $\vec{E}$ es conservativo, lo que matemáticamente implica que su rotacional es nulo:

$$\vec{\nabla}\times\vec{E}=\vec{0}$$

Esta propiedad nos permite definir un campo escalar $V$, conocido como potencial electrostático, cuya relación directa con el campo eléctrico está dada por:

$$\vec{E}=-\vec{\nabla}V$$

Para calcular la diferencia de potencial $\Delta V$ entre un punto inicial $i$ y un punto final $f$, integramos el campo eléctrico a lo largo de una trayectoria:

$$\Delta V=-\int_{i}^{f}\vec{E}\cdot d\vec{l}$$

## Campo eléctrico conservativo

Una de las ideas más contraintuitivas cuando comenzamos a estudiar el electromagnetismo es pensar que, al mover una carga entre dos puntos, el camino más largo o más sinuoso requerirá "gastar" más energía. Nuestra experiencia cotidiana con la fricción nos dice que la ruta importa. Sin embargo, en la electrostática, esto es estrictamente falso.

El campo eléctrico estático es un **campo conservativo**. Esto significa que el trabajo realizado por la fuerza eléctrica para mover una carga de prueba desde un punto inicial $i$ hasta un punto final $f$ es **independiente de la trayectoria**.

Matemáticamente, el trabajo $W$ por unidad de carga (medido en J/C o Voltios) se define a través de la integral de línea del producto punto entre el campo eléctrico $\vec{E}$ y el diferencial de desplazamiento $d\vec{l}$:

$$\frac{W}{q}=\int_{i}^{f}\vec{E}\cdot d\vec{l}=V_i-V_f$$

Como el campo eléctrico deriva de un potencial escalar ($\vec{E}=-\vec{\nabla} V$), esta integral depende únicamente de la diferencia de potencial entre los puntos extremos, sin importar qué camino se tome para llegar de uno a otro.

En nuestra primera simulación interactiva, exploramos el campo generado por una única carga puntual $+Q$. Las líneas de campo aquí divergen radialmente desde el origen. 

<iframe src="_static/potencial_trayectoria1.html" width="100%" height="600" style="border:none;"></iframe>

Observa atentamente las cuatro trayectorias. A medida que avanza la animación, notarás detalles clave sobre el diferencial de trabajo ($dW$):
* **A favor y en contra:** Cuando una trayectoria avanza en la misma dirección que las flechas del campo, el trabajo es positivo (se suma energía). Si la curva hace un bucle y viaja "contra la corriente", el producto punto $\vec{E}\cdot d\vec{l}$ se vuelve negativo, restando energía al acumulado.
* **Trabajo nulo en el arco:** Presta especial atención a la trayectoria del "Arco" (color morado). En la sección donde viaja trazando un círculo perfecto alrededor de la carga, se está moviendo de forma perpendicular a las líneas de campo radiales ($\vec{E}\perp d\vec{l}$). En esa zona geométrica, el diferencial es exactamente cero ($dW=0$). Se está moviendo sobre una **superficie equipotencial**.

Al finalizar, ya sea que la partícula haya viajado en línea recta, haya dado vueltas en un bucle o haya tomado un desvío diagonal, el trabajo total integrado será exactamente el mismo para las cuatro rutas.


## Simulación: Barra cargada y geometrías complejas

Es fácil pensar que la independencia de la trayectoria es solo una "suerte" matemática que ocurre por la simetría perfecta de una carga puntual. La segunda simulación está diseñada para romper ese mito.

<iframe src="_static/potencial_trayectoria2.html" width="100%" height="600" style="border:none;"></iframe>

Aquí, la fuente es una distribución continua: una barra horizontal de longitud $L$ con densidad de carga lineal uniforme ($+\lambda$). El campo eléctrico resultante es mucho más complejo. Ya no es puramente radial; cerca del centro las flechas apuntan hacia arriba, pero hacia los extremos se curvan violentamente debido a los efectos de borde. 

A pesar de esta geometría vectorial asimétrica e intrincada, el principio físico se mantiene inquebrantable. Las cuatro partículas navegan por encima de la barra, cruzando el campo en ángulos totalmente distintos, pero al aterrizar en el punto final $f$, el trabajo acumulado vuelve a coincidir a la perfección. La naturaleza conservativa no depende de la forma de la fuente, sino que es una propiedad fundamental del propio campo electrostático.


La simulación **"Cargas y Campos"** de PhET es una herramienta excelente para visualizar el campo eléctrico y el potencial electrostático. A continuación, se detallan los pasos para modelar y observar las superficies equipotenciales tanto de una carga puntual como de una distribución lineal de carga (barra uniforme).

### Configuración Inicial
1. Ingresa a la simulación.
2. En el menú lateral derecho, activa las siguientes opciones para facilitar la visualización y medición:
   * **Campo eléctrico:** Para visualizar las líneas de campo y su dirección.
   * **Grilla:** Para alinear las cargas con precisión y establecer referencias espaciales.
   * **Valores:** Para obtener lecturas numéricas del potencial eléctrico en voltios (**V**).

### Caso 1: Superficies Equipotenciales de una Carga Puntual
1. **Posicionamiento:** Arrastra una carga positiva de **+1 nC** desde el panel inferior hacia el centro de la grilla.
2. **Observación del campo:** Nota que las líneas del campo eléctrico $\vec{E}$ divergen con simetría radial desde la carga.
3. **Medición del potencial:** Selecciona la herramienta de medición de equipotenciales (el sensor azul con forma de cruz) y ubícala a una distancia específica de la carga (por ejemplo, a dos cuadrículas de distancia).
4. **Trazado de la superficie:** Haz clic en el ícono del **lápiz verde** adjunto al sensor. Esto trazará la superficie equipotencial correspondiente a ese voltaje, que para una carga puntual será una circunferencia perfecta.
5. **Mapeo del espacio:** Aleja el sensor radialmente a nuevas distancias y repite el trazado. Observarás que las superficies equipotenciales forman una familia de circunferencias concéntricas.

### Caso 2: Superficies Equipotenciales de una Barra Uniforme
Dado que la simulación opera con cargas puntuales, aproximaremos una distribución continua lineal agrupando múltiples cargas individuales.

1. **Limpieza del área:** Utiliza el botón de reinicio (flecha circular naranja en la esquina inferior derecha) para despejar el área de trabajo.
2. **Construcción de la distribución lineal:**
   * Coloca una carga de **+1 nC** en una intersección de la grilla.
   * Arrastra cargas adicionales de **+1 nC** y ubícalas contiguas a la primera, formando una línea recta horizontal ininterrumpida. Utiliza entre 8 y 10 cargas para obtener un buen modelo de la barra.


3. **Análisis del campo:** Observa que en la región central de la barra, las líneas de campo son aproximadamente paralelas y perpendiculares a la distribución. En los extremos, las líneas se curvan (efectos de borde).
4. **Trazado de las superficies:** Utiliza nuevamente el sensor de equipotenciales y ubícalo cerca del centro de la barra. Al trazar la línea con el lápiz, notarás una geometría ovalada o capsular que copia la simetría plana en el centro y se curva en los extremos.
5. **Evolución con la distancia:** Traza nuevas superficies alejándote progresivamente de la barra. Notarás un comportamiento dual:
   * **Zona cercana (campo próximo):** Las superficies son casi paralelas a la barra.
   * **Zona lejana (campo lejano):** Las superficies tienden a volverse circulares, demostrando que a distancias mucho mayores que la longitud de la barra, el sistema se aproxima al comportamiento de una carga puntual equivalente.

> **Nota sobre el Trabajo Eléctrico:** Si una carga de prueba se desplazara exactamente sobre cualquiera de estas trayectorias equipotenciales trazadas, el trabajo realizado por el campo eléctrico sería nulo ($W = 0$). Esto ocurre porque el desplazamiento diferencial $d\vec{l}$ es en todo momento perpendicular a las líneas de campo eléctrico $\vec{E}$, haciendo que el producto escalar $\vec{E} \cdot d\vec{l}$ sea cero.
