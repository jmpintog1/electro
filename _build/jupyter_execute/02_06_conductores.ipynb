# 2.6 Conductores y Aislantes

Para entender cómo se comporta el campo eléctrico en la materia mediante la Ley de Gauss, es fundamental diferenciar entre materiales conductores y no conductores (aislantes).

## Propiedades de los Conductores

* En los materiales conductores, las cargas eléctricas se mueven libremente en su interior.
* Debido a la repulsión mutua y a esta libertad de movimiento, cuando un conductor alcanza el equilibrio electrostático, las cargas se distribuyen exclusivamente en su superficie.

### Campo eléctrico dentro de la esfera ($r < R$):

Dado que toda la carga se aloja en la superficie del conductor, cualquier superficie gaussiana trazada en su interior no encerrará carga alguna.



```{image} images_tikz/02_07.svg
:alt: Esfera conductora dentro
:width: 40%
:align: center



$$\oint\vec{E}_{\text{I}}\cdot d\vec{A}=\frac{q_{\text{enc}}}{\epsilon_{0}}$$

Como $q_{\text{enc}}=0$, concluimos de inmediato que el campo eléctrico en el interior es nulo:

$$\vec{E}_{\text{I}}=\vec{0}$$

### Campo eléctrico fuera de la esfera ($r > R$):

Para una superficie gaussiana esférica de radio $r$ mayor al radio de la esfera ($r > R$), la carga encerrada es simplemente la carga total $Q$.



```{image} images_tikz/02_08.svg
:alt: Esfera conductora fuera
:width: 40%
:align: center



$$\oint\vec{E}_{\text{II}}\cdot d\vec{A}=\frac{q_{\text{enc}}}{\epsilon_{0}}$$

El flujo a través de esta superficie es:

$$\oint\vec{E}_{\text{II}}\cdot d\vec{A}=E_{\text{II}} \, 4\pi r^{2}$$

Igualando y despejando obtenemos el campo exterior, el cual decae proporcionalmente a $\frac{1}{r^{2}}$:

$$\begin{align}
E_{\text{II}} \, 4\pi r^{2}&=\frac{Q}{\epsilon_{0}} \\
\vec{E}_{\text{II}}&=\frac{Q}{4\pi\epsilon_{0}r^{2}}\hat{r}
\end{align}$$

## Esfera Aislante o No Conductora

A diferencia de un conductor, en una esfera aislante de radio $R$ la carga total $Q$ puede estar distribuida de manera uniforme por todo su volumen. Su densidad volumétrica de carga es constante: $\rho=\frac{Q}{\frac{4}{3}\pi R^{3}}$.



```{image} images_tikz/02_10.svg
:alt: Esfera aislante
:width: 40%
:align: center



### Campo eléctrico dentro de la esfera ($r < R$):

Al trazar una superficie gaussiana en su interior, la carga encerrada dependerá del volumen de esa superficie gaussiana, siendo una fracción de la carga total: $q_{\text{enc}}=\rho\frac{4}{3}\pi r^{3}$.



```{image} images_tikz/02_11.svg
:alt: Esfera aislante dentro
:width: 40%
:align: center



Aplicando la Ley de Gauss:

$$\begin{align}
\oint\vec{E}_{\text{I}}\cdot d\vec{A}&=\frac{q_{\text{enc}}}{\epsilon_{0}} \\
E_{\text{I}}\cdot 4\pi r^{2}&=\frac{Q\frac{r^{3}}{R^{3}}}{\epsilon_{0}}
\end{align}$$

Al despejar, notamos que en el interior de un aislante uniforme el campo eléctrico crece de forma directamente proporcional al radio $r$:

$$E_{\text{I}}=\frac{Qr}{4\pi\epsilon_{0}R^{3}}$$

Expresado en función de la densidad volumétrica $\rho$:

$$\begin{align}
E_{\text{I}}&=\frac{Q}{\frac{4}{3}\pi R^{3}}\frac{r}{3\epsilon_{0}} \\
E_{\text{I}}&=\rho\frac{r}{3\epsilon_{0}} \\
\vec{E}_{\text{I}}&=\frac{\rho r}{3\epsilon_{0}}\hat{r} \\
\vec{E}_{\text{I}}&=\frac{\rho}{3\epsilon_{0}}\vec{r}
\end{align}$$


### Campo eléctrico fuera de la esfera ($r > R$):

Desde el exterior, la esfera aislante se comporta matemáticamente igual que la esfera conductora o que una carga puntual. Toda la carga $Q$ queda encerrada por la superficie gaussiana.



```{image} images_tikz/02_12.svg
:alt: Esfera aislante fuera
:width: 40%
:align: center



$$\begin{align}
\oint\vec{E}_{\text{II}}\cdot d\vec{A}&=\frac{q_{\text{enc}}}{\epsilon_{0}} \\
E_{\text{II}} \, 4\pi r^{2}&=\frac{Q}{\epsilon_{0}} \\
\vec{E}_{\text{II}}&=\frac{Q}{4\pi\epsilon_{0}r^{2}}\hat{r}
\end{align}$$
