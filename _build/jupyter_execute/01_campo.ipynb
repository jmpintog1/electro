# Capítulo 1: Carga Eléctrica y Campo Eléctrico

La **carga eléctrica** ($q$) es una propiedad intrínseca de la materia que determina la intensidad de las interacciones electromagnéticas.

### Naturaleza de la Carga
* **Protones:** Poseen carga positiva ($q > 0$).
* **Electrones:** Poseen carga negativa ($q < 0$).
* **Neutrones:** Son eléctricamente neutros ($q = 0$).

### Unidades y Cuantización
La unidad de medida de la carga eléctrica en el Sistema Internacional es el **Coulomb** ($C$).

La **carga elemental** representa la magnitud de carga más pequeña que puede existir de forma aislada:
$e = 1{,}6 \times 10^{-19} \text{ C}$

De acuerdo a esto, las partículas fundamentales se definen como:
* **Protones:** $q_p = +e$
* **Electrones:** $q_e = -e$

> **Nota:** Un cuerpo adquiere carga neta al ceder o ganar electrones. Si pierde electrones, resulta con carga positiva; si los gana, su carga será negativa.

# Ley de Coulomb

La interacción entre cargas eléctricas se rige por los siguientes principios fundamentales:

* **Tipos de carga:** Existen dos tipos de cargas, positivas y negativas.
* **Interacción a distancia:** Dos cargas interactúan con una fuerza cuya magnitud es inversamente proporcional al cuadrado de la distancia $d$ que las separa.
* **Proporcionalidad:** Esta fuerza es directamente proporcional al producto de las magnitudes de las cargas ($q_1, q_2$).

## Formulación Matemática (Escalar)

La magnitud de la fuerza de Coulomb se define como:

$$F = k \frac{q_1 q_2}{d^2} = \frac{1}{4\pi\epsilon_0} \frac{q_1 q_2}{d^2}$$

Donde las constantes fundamentales son:
* **Constante de Coulomb ($k$):** $k =  8{,}98755 \times 10^9 \ \frac{\text{N}\cdot\text{m}^2}{\text{C}^2} \approx 9 \times 10^9 \ \frac{\text{N}\cdot\text{m}^2}{\text{C}^2}$
* **Permitividad del vacío ($\epsilon_0$):** $\epsilon_0 = 8{,}854 \times 10^{-12} \ \frac{\text{C}^2}{\text{N}\cdot\text{m}^2}$

---

## Formulación Vectorial

Para un análisis preciso en el espacio, definimos la fuerza que la carga $2$ ejerce sobre la carga $1$ ($\vec{F}_{12}$) utilizando vectores de posición $\vec{r}_1$ y $\vec{r}_2$:

```{image} images_tikz/01_04.svg
:alt: Coulomb
:width: 60%
:align: center

$$
\begin{aligned}
\vec{F}_{12} &= \frac{1}{4\pi\epsilon_0} \frac{q_1 q_2}{d^2} \hat{u} \\
&= \frac{1}{4\pi\epsilon_0} \frac{q_1 q_2}{|\vec{r}_1 - \vec{r}_2|^2} \frac{(\vec{r}_1 - \vec{r}_2)}{|\vec{r}_1 - \vec{r}_2|} \\
\vec{F}_{12} &= \frac{1}{4\pi\epsilon_0} q_1 q_2 \frac{(\vec{r}_1 - \vec{r}_2)}{|\vec{r}_1 - \vec{r}_2|^3}
\end{aligned}
$$

---

## Principio de Superposición (Cargas Discretas)

Si tenemos un sistema de $N$ cargas puntuales, la fuerza total resultante sobre una carga $q_1$ es la suma vectorial de las fuerzas individuales ejercidas por el resto de las cargas:

$$\vec{F}_1 = \vec{F}_{12} + \vec{F}_{13} + \dots + \vec{F}_{1N}$$

En términos generales:

$$\vec{F}_1 = \frac{q_1}{4\pi\epsilon_0} \sum_{j=2}^{N} q_j \frac{\vec{r}_1 - \vec{r}_j}{|\vec{r}_1 - \vec{r}_j|^3}$$

> **Nota:** Este análisis es válido estrictamente para distribuciones de **cargas puntuales**.

### **Problema: Equilibrio de fuerzas eléctricas**

**Enunciado:**

Considere dos cargas eléctricas puntuales, $q_1 = +q$ y $q_2 = +2q$, separadas por una distancia $d$. 

Se desea encontrar la posición exacta en la que una tercera carga $q_3$ debe ser colocada a lo largo de la línea que une $q_1$ y $q_2$, de modo que la **fuerza eléctrica neta** sobre $q_3$ sea cero. Es decir, que las fuerzas ejercidas por $q_1$ y $q_2$ sobre $q_3$ se cancelen mutuamente.

````python
import numpy as np
import matplotlib.pyplot as plt
from ipywidgets import interact, FloatSlider

def visualizar_equilibrio(x3):
    # Parámetros fijos
    d = 1.0    # Distancia entre q1 y q2
    q = 1e-6   # Magnitud base de carga (1 uC)
    q1 = q
    q2 = 2*q
    q3 = q     # La carga de prueba
    k = 8.987e9
    
    # Posiciones
    x1, x2 = 0, d
    
    # Cálculo de magnitudes de fuerza (solo en el eje X)
    # F13: Fuerza de q1 sobre q3
    r13 = abs(x3 - x1)
    if r13 < 0.05: r13 = 0.05 # Evitar división por cero
    F13 = k * q1 * q3 / (r13**2)
    if x3 < x1: F13 = -F13 # Dirección
    
    # F23: Fuerza de q2 sobre q3
    r23 = abs(x3 - x2)
    if r23 < 0.05: r23 = 0.05
    F23 = k * q2 * q3 / (r23**2)
    if x3 > x2: F23 = F23
    else: F23 = -F23
    
    F_neta = F13 + F23

    # Graficación
    plt.figure(figsize=(10, 4))
    plt.axhline(0, color='black', lw=1, ls='--')
    
    # Dibujar cargas
    plt.scatter([x1, x2], [0, 0], c=['blue', 'red'], s=200, label='Cargas fijas ($q_1, q_2$)')
    plt.scatter([x3], [0], c='green', s=150, label='Carga de prueba ($q_3$)')
    
    # Dibujar vectores de fuerza
    scale = 0.1 / max(abs(F13), abs(F23), 1e-9) # Escalar para visualización
    plt.arrow(x3, 0, F13*scale, 0, head_width=0.05, head_length=0.03, fc='blue', ec='blue', label='$\vec{F}_{13}$')
    plt.arrow(x3, 0, F23*scale, 0, head_width=0.05, head_length=0.03, fc='red', ec='red', label='$\vec{F}_{23}$')
    
    plt.title(f'Posición $x_3$ = {x3:.3f} m | Fuerza Neta = {F_neta:.3e} N')
    plt.xlim(-0.5, 1.5)
    plt.ylim(-0.5, 0.5)
    plt.legend(loc='upper right')
    plt.grid(True, alpha=0.3)
    plt.show()

# Crear el deslizador interactivo
interact(visualizar_equilibrio, 
         x3=FloatSlider(value=0.5, min=-0.2, max=1.2, step=0.001, description='Posición $q_3$:'))
````
