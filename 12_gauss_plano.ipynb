# Simetría Plana

Imagina una lámina plana, infinita y muy delgada, con una densidad de carga superficial uniforme $\sigma$. Por simetría, sabemos que el campo eléctrico $\vec{E}$ debe apuntar perpendicularmente alejándose de la lámina (si la carga es positiva) en ambos lados. Además, como la lámina es infinita, el campo no puede "inclinarse" hacia ningún lado, y su magnitud debe ser constante a una distancia dada.

¿Qué superficie gaussiana imaginaria debemos elegir para calcular este campo?

Aquí es donde la Ley de Gauss nos permite ser creativos. Necesitamos una superficie cerrada donde el producto punto $\vec{E} \cdot d\vec{A}$ sea fácil de evaluar.

## Explorando Opciones en 3D

### Análisis de las Superficies

Al girar la simulación, deberías notar lo siguiente:

#### 1. El Cilindro (La "Caja de Pastillas" clásica)
Esta es una elección excelente. El cilindro se coloca atravesando la lámina, con sus tapas paralelas a ella.
* **En las caras laterales (curvas):** El vector de área $d\vec{A}$ apunta radialmente hacia afuera, mientras que el campo $\vec{E}$ apunta hacia arriba o hacia abajo. ¡Son perpendiculares! ($\theta = 90^\circ$). Por lo tanto, el flujo a través de los lados es **cero** ($\vec{E} \cdot d\vec{A} = 0$).
* **En las tapas (planas):** El vector $d\vec{A}$ y el campo $\vec{E}$ son perfectamente paralelos ($\theta = 0^\circ$). El flujo es simplemente $E \cdot A_{tapa}$.

#### 2. La Esfera (Inadecuada)
Aunque funcionó perfecto para la carga puntual, aquí es un desastre matemático.
* Los vectores normales $d\vec{A}$ apuntan radialmente hacia afuera desde el centro de la esfera.
* El campo $\vec{E}$ siempre apunta verticalmente.
* El ángulo entre ellos cambia constantemente en cada punto de la superficie esférica. Calcular la integral $\oint \vec{E} \cdot d\vec{A}$ sería una pesadilla.

#### 3. El Paralelepípedo (¡También funciona!)
¡Sorpresa! El cilindro no es la única opción válida. Una caja rectangular (paralelepípedo) funciona exactamente igual de bien.
* **En las cuatro caras laterales:** $d\vec{A}$ es perpendicular a $\vec{E}$. El flujo es **cero**.
* **En las dos tapas (superior e inferior):** $d\vec{A}$ es paralelo a $\vec{E}$. El flujo es $E \cdot A_{tapa}$.

### Conclusión Matemática

Tanto para el cilindro como para la caja, el flujo total $\Phi_E$ se reduce únicamente a lo que cruza las dos tapas (una arriba y otra abajo):

$$\Phi_E = \oint \vec{E} \cdot d\vec{A} = \int_{tapa\_sup} E \, dA + \int_{tapa\_inf} E \, dA = 2\,E\,A$$

Aplicando la Ley de Gauss:

$$\Phi_E = \frac{q_{\text{enc}}}{\epsilon_0}$$
$$2EA = \frac{\sigma A}{\epsilon_0}$$

Despejando el campo eléctrico $E$, notamos que el área $A$ de las tapas se cancela:

$$E = \frac{\sigma}{2\epsilon_0}$$

**¡Un resultado profundo!** El campo eléctrico generado por una lámina plana infinita es constante. No depende de la distancia a la lámina.
