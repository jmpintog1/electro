# 4.5. Dieléctricos y Polarización

Un material dieléctrico es un aislante no conductor (como vidrio, papel o teflón). Al colocar un material dieléctrico entre las placas de un condensador, su comportamiento se rige por la respuesta de los átomos o moléculas del material al campo eléctrico.

## 4.5.1. Dipolos Eléctricos en un Campo Externo
Si analizamos una molécula polar (dipolo eléctrico con momento dipolar $\vec{p} = q\vec{a}$), al someterla a un campo eléctrico uniforme $\vec{E}$, experimenta fuerzas iguales y opuestas que generan un torque neto:

$$\vec{\tau}_{\text{neto}} = \vec{p} \times \vec{E}$$

Este torque tiende a alinear los dipolos del material con la dirección del campo eléctrico externo. 

## 4.5.2. Vector Polarización y Ley de Gauss
Definimos el **vector polarización** $\vec{P}$ como el momento dipolar por unidad de volumen:

$$\vec{P} = \frac{d\vec{p}}{d\mathcal{V}}$$

El alineamiento de los dipolos genera cargas inducidas de polarización ($Q_p$) en las superficies del material. Para considerar estos efectos en medios materiales, la Ley de Gauss se generaliza introduciendo el vector **Desplazamiento Eléctrico** $\vec{D}$:

$$\oint \vec{D} \cdot d\vec{S} = Q_{\text{libre}}$$
Donde $\vec{D} = \epsilon_0\vec{E} + \vec{P}$.

## 4.5.3. Materiales Dieléctricos Lineales
Para la mayoría de los materiales isotrópicos a campos no muy intensos, la polarización es proporcional al campo eléctrico:

$$\vec{P} = \chi \vec{E}$$

Donde $\chi$ es la susceptibilidad eléctrica. En este caso, el desplazamiento eléctrico se simplifica a:

$$\vec{D} = (\epsilon_0 + \chi)\vec{E} = \epsilon\vec{E}$$

Definimos la **constante dieléctrica** $\kappa$ (kappa) del material como la razón adimensional:

$$\kappa = \frac{\epsilon}{\epsilon_0} \geq 1$$

# 4.6. Efecto de los Dieléctricos en la Capacitancia

Al insertar un dieléctrico de constante $\kappa$ llenando completamente el espacio entre las placas de un condensador aislado (carga constante), el campo eléctrico neto disminuye debido al campo inducido opuesto por los dipolos ($E = E_0/\kappa$). 

Como $\Delta V = Ed$, el voltaje también disminuye, lo que resulta en un **aumento de la capacitancia**:

$$C = \frac{Q}{\Delta V} = \frac{Q}{E_0 d / \kappa} = \kappa \frac{Q}{\Delta V_0} = \kappa C_0$$

La capacitancia de cualquier condensador aumenta en un factor $\kappa$ cuando el espacio entre sus conductores se llena completamente con un dieléctrico. Para un condensador de placas paralelas, su nueva capacitancia será $C = \frac{\kappa \epsilon_0 A}{d}$.
